## 1. ¿Qué es la Inteligencia Artificial?

La **Inteligencia Artificial** (IA) es una rama de la informática que busca crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana.

### Definiciones clave

| Término | Definición simple |
|---|---|
| **IA** | Sistemas que imitan capacidades cognitivas humanas |
| **Machine Learning** | Subconjunto de IA: el sistema *aprende* de datos sin ser programado explícitamente |
| **Deep Learning** | Subconjunto de ML: usa redes neuronales con muchas capas |
| **Modelo** | El resultado del entrenamiento: una función matemática que hace predicciones |

### La relación entre conceptos

```
┌─────────────────────────────────────┐
│     Inteligencia Artificial         │
│  ┌───────────────────────────────┐  │
│  │     Machine Learning          │  │
│  │  ┌─────────────────────────┐  │  │
│  │  │    Deep Learning        │  │  │
│  │  │  ┌───────────────────┐  │  │  │
│  │  │  │  LLMs / GenAI     │  │  │  │
│  │  │  └───────────────────┘  │  │  │
│  │  └─────────────────────────┘  │  │
│  └───────────────────────────────┘  │
└─────────────────────────────────────┘
```

> 💡 **Clave:** No toda IA es Machine Learning. Sistemas basados en reglas (como un filtro de spam simple con reglas fijas) también son IA.

````python
# Visualicemos la relación entre conceptos con un diagrama simple

conceptos = {
    "Inteligencia Artificial": [
        "Sistemas basados en reglas",
        "Machine Learning",
        "Robótica",
        "Sistemas expertos"
    ],
    "Machine Learning": [
        "Supervisado",
        "No supervisado",
        "Por refuerzo",
        "Deep Learning"
    ],
    "Deep Learning": [
        "Redes convolucionales (CNN)",
        "Redes recurrentes (RNN)",
        "Transformers",
        "LLMs / IA Generativa"
    ]
}

# Mostrar la jerarquía en texto
print("📊 Mapa de subcampos de la IA\n")
for area, subcampos in conceptos.items():
    print(f"🔷 {area}")
    for sub in subcampos:
        print(f"   ├─ {sub}")
    print()
````

## 2. Tipos de IA: estrecha, general y superinteligencia

| Tipo | Descripción | ¿Existe hoy? | Ejemplo |
|---|---|---|---|
| **IA Estrecha** (ANI) | Hace UNA tarea bien | ✅ Sí | ChatGPT, filtro de spam, recomendador de Netflix |
| **IA General** (AGI) | Hace CUALQUIER tarea cognitiva como un humano | ❌ No | Ciencia ficción (por ahora) |
| **Superinteligencia** (ASI) | Supera al humano en todo | ❌ No | Solo teoría |

### ¿Dónde estamos hoy?

Todo lo que usamos hoy — ChatGPT, Copilot, detección de fraude, autos autónomos — es **IA estrecha**.
Cada sistema está especializado en una tarea (o un conjunto acotado de tareas).

> 💡 **Reflexión:** Cuando decimos que GPT-4 "sabe de todo", en realidad es un sistema de IA estrecha muy capaz en tareas de lenguaje. No puede mover un brazo robótico ni hacer experimentos de laboratorio.

````python
# Ejercicio interactivo: clasificá cada sistema según su tipo de IA

sistemas = [
    {"nombre": "Filtro de spam de Gmail",       "tipo_correcto": "ANI"},
    {"nombre": "ChatGPT",                       "tipo_correcto": "ANI"},
    {"nombre": "HAL 9000 (2001: Odisea)",       "tipo_correcto": "AGI"},
    {"nombre": "Tesla Autopilot",               "tipo_correcto": "ANI"},
    {"nombre": "Skynet (Terminator)",           "tipo_correcto": "ASI"},
    {"nombre": "Recomendador de Spotify",       "tipo_correcto": "ANI"},
    {"nombre": "JARVIS (Iron Man)",             "tipo_correcto": "AGI"},
]

print("🧪 Clasificación de sistemas de IA\n")
print(f"{'Sistema':<35} {'Tipo real':<10} {'Ficción?'}")
print("─" * 60)

for s in sistemas:
    ficcion = "🎬 Ficción" if s["tipo_correcto"] in ["AGI", "ASI"] else "✅ Real"
    print(f"{s['nombre']:<35} {s['tipo_correcto']:<10} {ficcion}")

print("\n💡 Todos los sistemas REALES que usamos hoy son IA Estrecha (ANI).")
````

## 4. Ciclo de vida de un proyecto de IA

Un proyecto de IA no es "entrenar un modelo y listo". Tiene etapas claras, y la mayoría del tiempo se gasta en **preparar datos**, no en el modelo.

### Las 7 etapas

```
1. Definir el problema     → ¿Qué queremos predecir/generar/clasificar?
2. Recolectar datos        → ¿De dónde vienen? ¿Cuántos necesitamos?
3. Preparar datos          → Limpieza, transformación, features
4. Elegir modelo           → ¿Regresión lineal? ¿Red neuronal? ¿LLM?
5. Entrenar                → Ajustar parámetros con los datos
6. Evaluar                 → ¿Qué tan bien funciona? Métricas
7. Desplegar y monitorear  → Poner en producción y vigilar
```

### ¿Dónde se gasta el tiempo?

| Etapa | % del tiempo (aprox) |
|---|---|
| Definir problema | 5% |
| Recolectar datos | 20% |
| **Preparar datos** | **40%** |
| Modelar + entrenar | 15% |
| Evaluar | 10% |
| Desplegar | 10% |

> 💡 **Regla del 80/20:** El 80% del éxito de un proyecto de IA depende de la calidad de los datos, no de la sofisticación del modelo.

````python
import matplotlib.pyplot as plt

# Visualización: ¿dónde se gasta el tiempo en un proyecto de IA?
etapas = [
    "Definir\nproblema",
    "Recolectar\ndatos",
    "Preparar\ndatos",
    "Modelar +\nentrenar",
    "Evaluar",
    "Desplegar"
]
porcentajes = [5, 20, 40, 15, 10, 10]
colores = ["#a8d8ea", "#aa96da", "#fcbad3", "#ffffd2", "#a8e6cf", "#dcedc1"]

fig, ax = plt.subplots(figsize=(10, 5))
barras = ax.bar(etapas, porcentajes, color=colores, edgecolor="gray", linewidth=0.5)

# Resaltar la barra más grande
barras[2].set_edgecolor("red")
barras[2].set_linewidth(2)

for barra, pct in zip(barras, porcentajes):
    ax.text(
        barra.get_x() + barra.get_width() / 2,
        barra.get_height() + 1,
        f"{pct}%",
        ha="center", va="bottom", fontsize=12, fontweight="bold"
    )

ax.set_ylabel("% del tiempo", fontsize=12)
ax.set_title("¿Dónde se gasta el tiempo en un proyecto de IA?", fontsize=14, fontweight="bold")
ax.set_ylim(0, 50)
ax.spines[["top", "right"]].set_visible(False)
plt.tight_layout()
plt.show()

print("\n📌 La preparación de datos es siempre la etapa más costosa en tiempo.")
````

## 7. Tu primera predicción con código

Antes de meternos con modelos complejos, veamos la idea más simple de ML:
**encontrar un patrón en datos y usarlo para predecir**.

### El ejemplo más simple posible

Tenemos datos de horas de estudio y notas de un examen. Queremos predecir la nota de alguien que estudia 7 horas.

> 💡 Este es un problema de **regresión lineal**: encontrar la recta que mejor se ajusta a los datos.

````python
import matplotlib.pyplot as plt

# ── Datos de ejemplo ──────────────────────────────────────────
# Horas de estudio → Nota obtenida (escala 1-10)
horas = [1, 2, 3, 4, 5, 6, 8, 9, 10]
notas = [2, 3, 3.5, 5, 5.5, 6, 7.5, 8, 9]

print("📊 Dataset: Horas de estudio vs. Nota\n")
print(f"{'Horas':<10} {'Nota'}")
print("─" * 20)
for h, n in zip(horas, notas):
    print(f"{h:<10} {n}")

# ── Calcular la recta a mano (mínimos cuadrados) ──────────────
# y = m * x + b
n_datos = len(horas)
sum_x   = sum(horas)
sum_y   = sum(notas)
sum_xy  = sum(h * n for h, n in zip(horas, notas))
sum_x2  = sum(h ** 2 for h in horas)

# Fórmulas de regresión lineal simple
m = (n_datos * sum_xy - sum_x * sum_y) / (n_datos * sum_x2 - sum_x ** 2)  # pendiente
b = (sum_y - m * sum_x) / n_datos                                          # intercepto

print(f"\n🔢 Recta encontrada: nota = {m:.2f} × horas + {b:.2f}")

# ── Predicción ────────────────────────────────────────────────
horas_nueva = 7
prediccion  = m * horas_nueva + b
print(f"\n🎯 Predicción: si estudias {horas_nueva} horas → nota estimada = {prediccion:.1f}")

# ── Visualización ────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

# Datos reales
ax.scatter(horas, notas, color="#4d96ff", s=80, zorder=3, label="Datos reales")

# Recta del modelo
x_linea = [0, 11]
y_linea = [m * x + b for x in x_linea]
ax.plot(x_linea, y_linea, color="#ff6b6b", linewidth=2, label=f"Modelo: y = {m:.2f}x + {b:.2f}")

# Predicción
ax.scatter([horas_nueva], [prediccion], color="#ff6b6b", s=150, marker="★", zorder=4,
           label=f"Predicción ({horas_nueva}h → {prediccion:.1f})")

ax.set_xlabel("Horas de estudio", fontsize=12)
ax.set_ylabel("Nota", fontsize=12)
ax.set_title("Tu primera predicción con IA", fontsize=14, fontweight="bold")
ax.legend(fontsize=10)
ax.set_xlim(0, 11)
ax.set_ylim(0, 10.5)
ax.spines[["top", "right"]].set_visible(False)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\n💡 Esto es exactamente lo que hace un modelo de ML:")
print("   1. Recibe datos")
print("   2. Encuentra un patrón (la recta)")
print("   3. Usa el patrón para predecir datos nuevos")
````
