## 2. ¿Qué es la Inteligencia Artificial?

La **Inteligencia Artificial** (IA) es una rama de la informática que busca crear sistemas
capaces de realizar tareas que normalmente requieren inteligencia humana.

### ¿Qué tareas?

- Reconocer objetos en imágenes
- Entender y generar texto
- Jugar juegos mejor que humanos
- Conducir un vehículo de forma autónoma
- Predecir si un paciente tiene una enfermedad

### Definiciones clave

| Término | Definición simple |
|---|---|
| **IA** | Sistemas que imitan capacidades cognitivas humanas |
| **Machine Learning** | Subconjunto de IA: el sistema *aprende* de datos sin ser programado explícitamente |
| **Deep Learning** | Subconjunto de ML: usa redes neuronales con muchas capas |
| **Modelo** | El resultado del entrenamiento: una función matemática que hace predicciones |
| **Datos** | La materia prima: ejemplos de los que el modelo aprende |

->
> 💡 **Analogía:** Un modelo de IA es como un estudiante. Los datos son sus apuntes. El entrenamiento es estudiar. La evaluación es el examen.

## 3. La jerarquía: IA → ML → DL → LLMs

Estos términos se usan a veces como sinónimos, pero tienen relaciones de **contención** (uno dentro del otro).

```
┌─────────────────────────────────────┐
│     Inteligencia Artificial         │
│  ┌───────────────────────────────┐  │
│  │     Machine Learning          │  │
│  │  ┌─────────────────────────┐  │  │
│  │  │    Deep Learning        │  │  │
│  │  │  ┌───────────────────┐  │  │  │
│  │  │  │  LLMs / GenAI     │  │  │  │
│  │  │  └───────────────────┘  │  │  │
│  │  └─────────────────────────┘  │  │
│  └───────────────────────────────┘  │
└─────────────────────────────────────┘
```

### ¿Qué diferencia a cada nivel?

| Nivel | Clave | Ejemplo |
|---|---|---|
| **IA** | Cualquier sistema que imita inteligencia | Ajedrez con reglas, Siri, ChatGPT |
| **ML** | Aprende de datos automáticamente | Filtro de spam, recomendador |
| **DL** | Usa redes neuronales profundas | Reconocimiento de voz, imágenes |
| **LLMs** | Modelos de lenguaje gigantes tipo Transformer | GPT-4, Claude, LLaMA |

->
> 💡 **No toda IA es Machine Learning.** Un sistema de reglas tipo "si la temperatura > 38°C → alerta" es IA pero no ML.

````python
# Visualicemos la jerarquía como un diagrama interactivo

jerarquia = {
    "Inteligencia Artificial": [
        "Sistemas basados en reglas",
        "Sistemas expertos",
        "Machine Learning",
        "Robótica",
    ],
    "Machine Learning": [
        "Aprendizaje supervisado",
        "Aprendizaje no supervisado",
        "Aprendizaje por refuerzo",
        "Deep Learning",
    ],
    "Deep Learning": [
        "Redes convolucionales (CNN)",
        "Redes recurrentes (RNN)",
        "Transformers",
        "LLMs / IA Generativa",
    ]
}

print("📊 Jerarquía de subcampos de la IA\n")
for area, hijos in jerarquia.items():
    print(f"🔷 {area}")
    for i, hijo in enumerate(hijos):
        conector = "└─" if i == len(hijos) - 1 else "├─"
        print(f"   {conector} {hijo}")
    print()
````

## 4. Tipos de IA: estrecha, general y superinteligencia

| Tipo | Sigla | Descripción | ¿Existe hoy? |
|---|---|---|---|
| **IA Estrecha** | ANI | Hace UNA tarea (o pocas relacionadas) muy bien | ✅ Sí |
| **IA General** | AGI | Hace CUALQUIER tarea cognitiva como un humano | ❌ No |
| **Superinteligencia** | ASI | Supera al humano en absolutamente TODO | ❌ No |

### ¿Dónde estamos hoy?

Todo lo que usamos hoy — ChatGPT, Copilot, Siri, detección de fraude, autos semi-autónomos — es **IA estrecha**.

Un modelo de lenguaje como GPT-4 parece "saber de todo", pero en realidad es un sistema optimizado
para **predecir el siguiente token en una secuencia de texto**. No puede mover un brazo robótico,
hacer un experimento químico, ni tener experiencias subjetivas.

> 💡 **Reflexión:** ¿Creés que llegaremos a AGI? Es un debate abierto y apasionante en la comunidad científica.

````python
# Ejercicio interactivo: clasificá cada sistema según su tipo de IA

sistemas = [
    {"nombre": "Filtro de spam de Gmail",           "tipo": "ANI", "real": True},
    {"nombre": "ChatGPT / Claude",                  "tipo": "ANI", "real": True},
    {"nombre": "HAL 9000 (2001: Odisea del Espacio)","tipo": "AGI", "real": False},
    {"nombre": "Tesla Autopilot",                   "tipo": "ANI", "real": True},
    {"nombre": "Skynet (Terminator)",               "tipo": "ASI", "real": False},
    {"nombre": "Recomendador de Spotify",           "tipo": "ANI", "real": True},
    {"nombre": "JARVIS (Iron Man)",                 "tipo": "AGI", "real": False},
    {"nombre": "GitHub Copilot",                    "tipo": "ANI", "real": True},
    {"nombre": "Google Translate",                  "tipo": "ANI", "real": True},
    {"nombre": "Data (Star Trek)",                  "tipo": "AGI", "real": False},
]

print("🧪 Clasificación de sistemas de IA\n")
print(f"{'Sistema':<45} {'Tipo':<8} {'Estado'}")
print("─" * 70)

for s in sistemas:
    estado = "✅ Real" if s["real"] else "🎬 Ficción"
    print(f"{s['nombre']:<45} {s['tipo']:<8} {estado}")

print("\n📌 Patrón clave: todo lo que es REAL es ANI (IA estrecha).")
print("   AGI y ASI solo existen en la ciencia ficción... por ahora.")
````

## 5. Subcampos de la IA

La IA es un paraguas enorme que cubre muchas disciplinas diferentes.

### Los subcampos principales

| Subcampo | Qué hace | Ejemplo concreto |
|---|---|---|
| **NLP** | Entender y generar texto humano | ChatGPT, traductores, resúmenes |
| **Computer Vision** | Entender imágenes y video | Reconocimiento facial, diagnóstico médico |
| **Robótica** | Controlar agentes físicos | Drones autónomos, brazos industriales |
| **Sistemas de Recomendación** | Predecir preferencias | Netflix, Spotify, Amazon |
| **IA Generativa** | Crear contenido nuevo | DALL-E, Midjourney, Suno (música) |
| **Agentes Inteligentes** | Planificar y ejecutar tareas | Asistentes con herramientas |
| **Aprendizaje por Refuerzo** | Aprender por prueba y error | AlphaGo, robots que caminan |

### ¿En qué nos enfocamos en este curso?

Principalmente en **NLP** e **IA Generativa**, pero los conceptos de ML que aprenderás aplican a todos los subcampos.

> 💡 **Muchas aplicaciones reales combinan varios subcampos.** Un auto autónomo usa Computer Vision + IA de control + planificación.

````python
# Mini-quiz: asociar tarea → subcampo de IA

tareas = [
    {"tarea": "Traducir un email del español al inglés",           "subcampo": "NLP"},
    {"tarea": "Detectar un tumor en una radiografía",              "subcampo": "Computer Vision"},
    {"tarea": "Sugerirte una película en Netflix",                 "subcampo": "Sist. Recomendación"},
    {"tarea": "Generar una imagen a partir de texto",              "subcampo": "IA Generativa"},
    {"tarea": "Un robot que arma piezas en una fábrica",           "subcampo": "Robótica"},
    {"tarea": "Un bot que reserva vuelos consultando APIs",        "subcampo": "Agentes Inteligentes"},
    {"tarea": "Un programa que aprende a jugar Atari solo",        "subcampo": "Aprendizaje por Refuerzo"},
    {"tarea": "Resumir un documento de 50 páginas en 3 párrafos",  "subcampo": "NLP"},
    {"tarea": "Componer música original",                          "subcampo": "IA Generativa"},
    {"tarea": "Detectar si un email es phishing",                  "subcampo": "NLP + ML"},
]

print("🎯 ¿A qué subcampo pertenece cada tarea?\n")
print(f"{'Tarea':<55} {'Subcampo'}")
print("─" * 80)

for t in tareas:
    print(f"{t['tarea']:<55} → {t['subcampo']}")

print("\n📌 Muchas tareas reales combinan varios subcampos a la vez.")
````

## 6. Mapa conceptual visual

Construyamos un mapa que conecte todos los conceptos que vimos.

````python
# Mapa conceptual visual con matplotlib

fig, ax = plt.subplots(figsize=(14, 9))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect("equal")
ax.axis("off")

# Nodos: (x, y, ancho, alto, texto, color)
nodos = [
    (3.5, 9.0, 3.0, 0.7, "Inteligencia\nArtificial",        "#ff6b6b"),
    (0.5, 7.0, 2.2, 0.6, "Sistemas\nbasados en reglas",     "#c0c0c0"),
    (3.8, 7.0, 2.2, 0.6, "Machine\nLearning",               "#ffd93d"),
    (7.2, 7.0, 2.2, 0.6, "Robótica",                        "#c0c0c0"),
    (1.5, 5.0, 2.0, 0.6, "Supervisado",                     "#6bcb77"),
    (4.0, 5.0, 2.0, 0.6, "No supervisado",                  "#6bcb77"),
    (6.5, 5.0, 2.0, 0.6, "Deep Learning",                   "#4d96ff"),
    (1.5, 3.0, 2.0, 0.6, "Refuerzo",                        "#6bcb77"),
    (4.5, 3.0, 2.0, 0.6, "Transformers",                    "#9b59b6"),
    (7.5, 3.0, 2.0, 0.6, "CNN / RNN",                       "#9b59b6"),
    (4.5, 1.2, 2.5, 0.7, "LLMs / IA\nGenerativa",          "#e74c3c"),
]

for x, y, w, h, texto, color in nodos:
    rect = mpatches.FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.1",
        facecolor=color, edgecolor="#333", linewidth=1.5, alpha=0.85
    )
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, texto, ha="center", va="center",
            fontsize=9, fontweight="bold", color="#222")

# Flechas
flechas = [
    (5.0, 9.0, 1.6, 7.6),   # IA → Reglas
    (5.0, 9.0, 4.9, 7.6),   # IA → ML
    (5.0, 9.0, 8.3, 7.6),   # IA → Robótica
    (4.9, 7.0, 2.5, 5.6),   # ML → Supervisado
    (4.9, 7.0, 5.0, 5.6),   # ML → No supervisado
    (4.9, 7.0, 7.5, 5.6),   # ML → DL
    (4.9, 7.0, 2.5, 3.6),   # ML → Refuerzo
    (7.5, 5.0, 5.5, 3.6),   # DL → Transformers
    (7.5, 5.0, 8.5, 3.6),   # DL → CNN
    (5.5, 3.0, 5.75, 1.9),  # Transformers → LLMs
]

for x1, y1, x2, y2 in flechas:
    ax.annotate("", xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle="->", color="#555", lw=1.2))

ax.set_title("Mapa conceptual: Inteligencia Artificial y sus ramas",
             fontsize=14, fontweight="bold", pad=15)
plt.tight_layout()
plt.show()

print("\n📌 Este mapa es tu guía de referencia para todo el curso.")
print("   Cada rama la iremos profundizando clase a clase.")
````
