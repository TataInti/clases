## 1. La intuición: aprender de ejemplos

Pensá cómo aprende un niño a distinguir un perro de un gato:

1. Ve **muchos** ejemplos de cada uno.
2. Alguien le dice *"esto es un perro, esto es un gato"*.
3. Con el tiempo, **descubre patrones** (orejas, tamaño, ladrido).
4. Cuando ve un animal nuevo, **generaliza** lo aprendido.

El Machine Learning hace exactamente lo mismo, pero con **números**:

```
Datos históricos  ─►  Algoritmo  ─►  Modelo  ─►  Predicciones sobre datos nuevos
    (ejemplos)         (ajusta)     (reglas)
```

### Diferencia clave con la programación tradicional

| Programación tradicional | Machine Learning |
|---|---|
| **Humano escribe reglas** | **Máquina descubre reglas** |
| Datos + reglas → resultado | Datos + resultados → reglas |
| Determinista | Probabilístico |
| Ej: "Si temp > 38 → fiebre" | Ej: aprender de 10.000 historias clínicas |

> ⚠️ ML **no es magia**: necesita muchos ejemplos, datos de calidad y una tarea bien definida.

## 2. Los 3 paradigmas del Machine Learning

Según **qué información tiene el modelo** al aprender, hay tres grandes familias:

| Paradigma | ¿Qué recibe el modelo? | ¿Qué aprende? | Ejemplo típico |
|---|---|---|---|
| **Supervisado** | Ejemplos **con respuesta correcta** | A predecir la respuesta | Detectar spam, predecir precios |
| **No supervisado** | Ejemplos **sin respuesta** | Estructuras ocultas | Segmentar clientes, detectar anomalías |
| **Por refuerzo** | Un entorno con **recompensas** | Una estrategia óptima | Jugar ajedrez, conducción autónoma |

```
              ┌──────────────────────────────────────────┐
              │           MACHINE LEARNING               │
              └──────────────────────────────────────────┘
                   │              │              │
         ┌─────────▼──────┐  ┌────▼──────┐  ┌────▼────────┐
         │  Supervisado   │  │ No superv.│  │  Refuerzo   │
         │ (con etiqueta) │  │ (patrones)│  │ (recompensa)│
         └────────────────┘  └───────────┘  └─────────────┘
```

Veamos cada uno con un ejemplo visual.

## 3. Aprendizaje supervisado (el más común)

**Idea:** le mostramos al modelo muchos pares `(entrada, respuesta correcta)`
y aprende a predecir la respuesta cuando llega una entrada nueva.

### Dos sabores

| Tipo | La respuesta es... | Ejemplo |
|---|---|---|
| **Clasificación** | Una **categoría** | ¿spam o no spam? ¿benigno o maligno? |
| **Regresión** | Un **número** | ¿cuánto costará esta casa? ¿cuántas ventas habrá? |

Debajo vemos una **clasificación** en acción: el modelo aprende a dibujar una frontera
que separa dos grupos de puntos.


````python
# Demo visual: clasificación supervisada
# El modelo ve ejemplos etiquetados (azul / rojo) y aprende a separarlos

import matplotlib.pyplot as plt
import random
import math

random.seed(7)

# Generamos dos grupos sintéticos (sin sklearn, puro Python)
azules = [(random.gauss(2, 0.7), random.gauss(2, 0.7)) for _ in range(40)]
rojos  = [(random.gauss(5, 0.7), random.gauss(5, 0.7)) for _ in range(40)]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Izquierda: los datos "crudos" como los ve el modelo
axes[0].scatter(*zip(*azules), c="#3498db", s=60, edgecolors="#222", label="Clase A")
axes[0].scatter(*zip(*rojos),  c="#e74c3c", s=60, edgecolors="#222", label="Clase B")
axes[0].set_title("Datos de entrenamiento\n(cada punto tiene una etiqueta)")
axes[0].legend(); axes[0].grid(alpha=0.3)

# Derecha: lo mismo + la "frontera" que el modelo aprende
axes[1].scatter(*zip(*azules), c="#3498db", s=60, edgecolors="#222")
axes[1].scatter(*zip(*rojos),  c="#e74c3c", s=60, edgecolors="#222")
# Línea separadora aprendida (conceptualmente: y = -x + 7)
xs = [0, 7]; ys = [7, 0]
axes[1].plot(xs, ys, "k--", linewidth=2, label="Frontera aprendida")
axes[1].fill_between(xs, ys, 0, color="#3498db", alpha=0.08)
axes[1].fill_between(xs, ys, 8, color="#e74c3c", alpha=0.08)
axes[1].set_title("El modelo aprende una frontera\nque separa las clases")
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
````

> 💡 **Lo importante:** el modelo no "memorizó" los puntos. Aprendió una **regla general**
> (la línea) que le permite clasificar **puntos nuevos** que nunca vio.
>
> Cuando esa línea no existe o es muy complicada, usamos modelos más poderosos
> (árboles, redes neuronales), pero **la idea es la misma**.

## 4. Aprendizaje no supervisado

**Idea:** el modelo recibe datos **sin etiqueta** y busca estructura por sí mismo.
Nadie le dice "esto es A, esto es B" — tiene que descubrirlo.

### Tareas típicas

| Tarea | ¿Qué hace? | Ejemplo real |
|---|---|---|
| **Clustering** | Agrupa ejemplos parecidos | Segmentar clientes por comportamiento |
| **Reducción de dimensiones** | Resume muchas variables en pocas | Visualizar datos complejos |
| **Detección de anomalías** | Encuentra lo que se sale del patrón | Fraude bancario, fallas industriales |


````python
# Demo visual: clustering (no supervisado)
# El modelo ve los puntos SIN colores y tiene que descubrir los grupos

random.seed(7)
grupo1 = [(random.gauss(2, 0.6), random.gauss(2, 0.6)) for _ in range(30)]
grupo2 = [(random.gauss(6, 0.6), random.gauss(2, 0.6)) for _ in range(30)]
grupo3 = [(random.gauss(4, 0.6), random.gauss(6, 0.6)) for _ in range(30)]
todos  = grupo1 + grupo2 + grupo3

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Izquierda: lo que ve el modelo (todo gris, sin etiquetas)
axes[0].scatter(*zip(*todos), c="#7f8c8d", s=60, edgecolors="#222")
axes[0].set_title("Lo que recibe el modelo:\ndatos sin etiqueta")
axes[0].grid(alpha=0.3)

# Derecha: lo que el algoritmo "descubre"
for puntos, color, nombre in [(grupo1, "#3498db", "Cluster 1"),
                               (grupo2, "#e74c3c", "Cluster 2"),
                               (grupo3, "#2ecc71", "Cluster 3")]:
    axes[1].scatter(*zip(*puntos), c=color, s=60, edgecolors="#222", label=nombre)
axes[1].set_title("Lo que descubre el algoritmo:\n3 grupos naturales")
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
````

### ⚠️ Ojo con las conclusiones

El modelo encontró 3 grupos, **pero no sabe qué significan**.
Es el humano quien interpreta:

> *"El cluster 1 parece clientes ahorrativos, el 2 parece impulsivos, el 3 parece VIPs..."*

Esto es potente pero peligroso: **podemos leer lo que queremos ver** en los grupos.
Un buen analista valida los clusters con datos externos antes de actuar.

## 5. Aprendizaje por refuerzo (RL)

**Idea:** un **agente** interactúa con un **entorno**, toma acciones y recibe **recompensas**
o **castigos**. Con el tiempo aprende una **estrategia** (política) que maximiza la recompensa.

```
           ┌──────────┐      acción      ┌──────────┐
           │  AGENTE  │ ───────────────► │ ENTORNO  │
           │ (modelo) │ ◄─────────────── │ (mundo)  │
           └──────────┘   estado +       └──────────┘
                           recompensa
```

### Ejemplos clásicos

| Dominio | Agente | Recompensa |
|---|---|---|
| Ajedrez / Go | AlphaZero | +1 al ganar, −1 al perder |
| Videojuegos | Agente DQN | Puntaje del juego |
| Robótica | Brazo robot | Llegar al objetivo sin chocar |
| LLMs (RLHF) | ChatGPT | Preferencia humana sobre respuestas |

> 💡 RL es más difícil de entrenar que supervisado, pero potentísimo cuando **no existe un
> conjunto de respuestas correctas** y sí un **objetivo medible**.

## 6. ¿Cuándo conviene usar ML?

ML **no siempre es la respuesta**. Antes de iniciar un proyecto, preguntate:

### ✅ Buenos candidatos para ML

- El problema tiene **patrones** pero **reglas explícitas son imposibles**
  (ej: detectar un tumor en una radiografía).
- Hay **mucha data histórica** disponible.
- El costo de un error es **tolerable** o mitigable.
- La tarea se repite millones de veces (vale la pena automatizar).

### ❌ Malos candidatos para ML

- El problema se resuelve con **una fórmula conocida** (no hace falta aprender nada).
- Hay **pocos datos** o los datos son **de mala calidad**.
- Se requiere **explicabilidad total y trazabilidad legal** (ej: decisiones judiciales) — usar ML es posible pero riesgoso.
- Los costos de error son **catastróficos e irreversibles** sin supervisión humana.

> 🎯 **Regla de oro:** si un experto humano puede hacer la tarea en **menos de un segundo**
> mirando los datos, y hay **muchos ejemplos**, ML probablemente funcione.
> (Frase atribuida a Andrew Ng.)

## 7. El ciclo de vida de un proyecto de IA

Un proyecto de ML **nunca es solo entrenar un modelo**. Tiene etapas:

```

  1. Definir          2. Obtener        3. Explorar y
     problema    ──►     datos     ──►   limpiar          
                                                          
                                                         ▼
  6. Monitorear   ◄── 5. Desplegar  ◄─── 4. Entrenar y
     en producción       el modelo         evaluar
        │                                     
        └──► 7. Reentrenar con datos nuevos (vuelve al 2)
        ***
```

| Etapa | ¿Quién suele encargarse? | % de tiempo típico |
|---|---|---|
| Definir problema | Producto + negocio | 5% |
| Obtener datos | Ing. de datos | 20% |
| Limpiar y explorar | Data scientist | **40%** ⚠️ |
| Entrenar y evaluar | Data scientist / ML eng. | 15% |
| Desplegar | ML engineer / DevOps | 10% |
| Monitorear | ML engineer + negocio | 10% |

> 📊 **Sorpresa común:** la mayor parte del tiempo NO se va entrenando modelos,
> sino **limpiando y entendiendo datos**. Por eso la clase 3 fue tan importante.


### 🎮 Ejercicio: jugá con el clasificador

Abajo tenés el mismo clasificador de antes, pero convertido en una **mini función interactiva**.
Cambiá los parámetros y observá qué pasa.

**Tu misión (tocá solo los valores marcados con `# 👈 TOCAR`):**

1. **Acercá los grupos** (`centro_A` y `centro_B` más cercanos). ¿La frontera los sigue separando bien? ¿Qué pasa cuando se solapan?
2. **Agrandá la dispersión** (`ruido` más alto). ¿Qué porcentaje de puntos quedan del lado equivocado?
3. **Achicá el dataset** (`n_puntos = 5`). Con pocos ejemplos, ¿la frontera sigue siendo confiable?
4. **Bonus:** agregá un **punto nuevo** (variable `punto_nuevo`) y fijate de qué lado cae.

> 🎯 Objetivo: entender **intuitivamente** que la calidad del modelo depende de:
> (a) cuán separables son las clases, (b) cuánto ruido hay, (c) cuántos ejemplos tenés.


````python
# 🎮 Clasificador de juguete — tocá los valores con 👈 y volvé a correr

# ── Parámetros que podés cambiar ──────────────────────────────
centro_A = (2, 2)      # 👈 TOCAR: centro del grupo azul
centro_B = (5, 5)      # 👈 TOCAR: centro del grupo rojo
ruido    = 0.7         # 👈 TOCAR: dispersión de los puntos (0.2 = compactos, 2.0 = caóticos)
n_puntos = 40          # 👈 TOCAR: cuántos puntos por clase
punto_nuevo = (3.5, 3.5)  # 👈 TOCAR: un punto nuevo a clasificar
# ──────────────────────────────────────────────────────────────

random.seed(7)
A = [(random.gauss(centro_A[0], ruido), random.gauss(centro_A[1], ruido)) for _ in range(n_puntos)]
B = [(random.gauss(centro_B[0], ruido), random.gauss(centro_B[1], ruido)) for _ in range(n_puntos)]

# "Entrenamiento" simplísimo: la frontera pasa por el punto medio entre los dos centros,
# perpendicular a la recta que los une. No hace falta entenderlo ahora.
mx, my = (centro_A[0] + centro_B[0]) / 2, (centro_A[1] + centro_B[1]) / 2
dx, dy = centro_B[0] - centro_A[0], centro_B[1] - centro_A[1]

def lado(p):
    """Devuelve 'A' o 'B' según de qué lado de la frontera cae el punto p."""
    return "A" if (p[0] - mx) * dx + (p[1] - my) * dy < 0 else "B"

# Evaluar cuántos puntos quedan bien clasificados
aciertos = sum(lado(p) == "A" for p in A) + sum(lado(p) == "B" for p in B)
total    = len(A) + len(B)

# ── Visualización ─────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(*zip(*A), c="#3498db", s=60, edgecolors="#222", label="Clase A")
ax.scatter(*zip(*B), c="#e74c3c", s=60, edgecolors="#222", label="Clase B")

# Dibujar la frontera aprendida
if dy != 0:
    xs = [mx - 5, mx + 5]
    ys = [my + (dx / dy) * (mx - x) for x in xs]
    ax.plot(xs, ys, "k--", linewidth=2, label="Frontera")

# Dibujar el punto nuevo con su predicción
pred = lado(punto_nuevo)
color_pred = "#3498db" if pred == "A" else "#e74c3c"
ax.scatter(*punto_nuevo, s=400, c=color_pred, edgecolors="black", linewidths=3,
           marker="*", label=f"Punto nuevo → Clase {pred}", zorder=10)

ax.set_title(f"Precisión: {aciertos}/{total} = {aciertos/total:.0%}", fontsize=13)
ax.legend(loc="best"); ax.grid(alpha=0.3)
ax.set_xlim(-2, 9); ax.set_ylim(-2, 9)
plt.tight_layout(); plt.show()

print(f"🎯 El punto {punto_nuevo} fue clasificado como: Clase {pred}")
print(f"📊 Precisión sobre los datos de entrenamiento: {aciertos/total:.1%}")
````
