## 1. De dónde viene el Machine Learning

La Inteligencia Artificial no empezó con ChatGPT ni con los modelos de lenguaje. Durante décadas, muchos sistemas inteligentes se construyeron con **reglas escritas por personas**: si pasa A, entonces hacer B.

Ese enfoque funciona bien cuando el problema es claro y estable. Por ejemplo:

| Problema | Regla posible |
|---|---|
| Aplicar un descuento | Si el cliente compra más de 10 unidades, descontar 15% |
| Validar un formulario | Si falta el DNI, mostrar error |
| Calcular un impuesto | Aplicar una fórmula definida por ley |

El límite aparece cuando las reglas son difíciles de escribir a mano. ¿Cómo describimos todas las formas posibles de detectar fraude? ¿Todas las formas en que una persona expresa una queja? ¿Todas las señales tempranas de abandono escolar? Ahí aparece Machine Learning.

**Machine Learning** es una forma de construir sistemas que aprenden patrones desde datos históricos. En vez de escribir todas las reglas, reunimos ejemplos y dejamos que un algoritmo encuentre regularidades.

| Enfoque basado en reglas | Enfoque con Machine Learning |
|---|---|
| Una persona escribe la lógica | El algoritmo aprende desde ejemplos |
| Sirve si las reglas son conocidas | Sirve si el patrón es complejo |
| Cambiar el sistema implica cambiar reglas | Cambiar el sistema implica mejorar datos, etiquetas o modelo |
| Es más transparente | Puede ser menos interpretable |

> Idea central: ML no reemplaza el razonamiento humano. Automatiza el reconocimiento de patrones cuando el problema es demasiado variable para resolverse con reglas simples.

## 2. IA, Machine Learning, Deep Learning y LLMs

Conviene ordenar algunos términos que suelen mezclarse.

| Concepto | Qué significa | Ejemplo |
|---|---|---|
| **Inteligencia Artificial (IA)** | Campo amplio: sistemas que realizan tareas asociadas a inteligencia humana | Un sistema experto, un chatbot, un recomendador |
| **Machine Learning (ML)** | Subcampo de IA donde el sistema aprende patrones desde datos | Predecir abandono de clientes |
| **Deep Learning** | Subcampo de ML basado en redes neuronales profundas | Reconocimiento de imágenes, voz, traducción |
| **Modelos de lenguaje / LLMs** | Modelos de deep learning entrenados con grandes volúmenes de texto | ChatGPT, Gemini, Claude, Llama |

Una forma simple de verlo:

```text
Inteligencia Artificial
└── Machine Learning
    └── Deep Learning
        └── Modelos de lenguaje grandes (LLMs)
```

En las clases anteriores usamos LLMs ya entrenados y aprendimos a orientarlos con prompts. En Machine Learning clásico, muchas veces construimos un modelo más pequeño y específico para una tarea concreta: clasificar consultas, estimar demanda, detectar anomalías, priorizar casos o recomendar acciones.

| Prompting con LLM | Machine Learning clásico |
|---|---|
| Partimos de un modelo general ya entrenado | Partimos de datos propios o de un dataset específico |
| Lo guiamos con instrucciones y ejemplos | Lo entrenamos con variables y etiquetas |
| Es flexible y conversacional | Es más acotado y medible para una tarea específica |
| Puede resolver tareas muy variadas | Suele resolver muy bien una tarea concreta |

No son enemigos. En sistemas reales pueden complementarse: un modelo de ML detecta casos relevantes y un LLM ayuda a explicar, resumir o redactar una respuesta.

## 3. Qué problemas resuelve Machine Learning

ML sirve cuando queremos usar datos del pasado para apoyar decisiones futuras. La pregunta no es “¿podemos usar IA?”, sino “¿hay un patrón aprendible en los datos?”.

| Tipo de problema | Pregunta típica | Ejemplo empresarial o público |
|---|---|---|
| **Clasificación** | ¿A qué categoría pertenece este caso? | ¿Esta consulta es un reclamo, una sugerencia o un pedido de ayuda? |
| **Regresión** | ¿Qué valor numérico esperamos? | ¿Cuánta demanda habrá la próxima semana? |
| **Ranking / priorización** | ¿Qué casos conviene atender primero? | ¿Qué expedientes tienen mayor urgencia? |
| **Detección de anomalías** | ¿Qué casos se salen de lo normal? | ¿Qué transacciones podrían ser fraudulentas? |
| **Recomendación** | ¿Qué opción es más relevante para esta persona? | ¿Qué trámite, producto o contenido sugerir? |
| **Segmentación** | ¿Qué grupos parecidos existen? | ¿Qué perfiles de usuarios tienen necesidades similares? |
| **Generación** | ¿Qué contenido nuevo puede producirse? | Borradores de respuestas, resúmenes, reportes |

Para decidir si ML tiene sentido, podemos usar esta guía:

| Situación | ¿ML ayuda? | Alternativa más simple |
|---|---|---|
| La regla es conocida y estable | Poco | Fórmulas, reglas, validaciones |
| Hay datos históricos de calidad | Sí | — |
| El patrón cambia con el tiempo | Sí, con monitoreo | Reglas actualizadas manualmente |
| Hay muy pocos casos | Con cuidado | Revisión humana, análisis descriptivo |
| El costo de error es alto | Sí, pero con supervisión | Decisión humana asistida |
| Se necesita explicar cada decisión | Depende del modelo | Reglas o modelos interpretables |

> Pregunta clave para cualquier proyecto: ¿qué decisión concreta queremos mejorar con datos?

## 4. Tipos de aprendizaje automático

No todo ML aprende de la misma manera. La diferencia principal está en el tipo de datos disponibles y en la forma de recibir feedback.

| Tipo de aprendizaje | Qué datos usa | Qué aprende | Ejemplo |
|---|---|---|---|
| **Supervisado** | Ejemplos con respuesta correcta | A predecir una etiqueta o un valor | Clasificar consultas, estimar ventas |
| **No supervisado** | Datos sin respuesta correcta | A encontrar grupos o patrones ocultos | Segmentar usuarios, detectar temas frecuentes |
| **Por refuerzo** | Acciones, recompensas y castigos | A elegir acciones que maximizan una recompensa | Optimizar rutas, juegos, robots |
| **Semi-supervisado** | Pocos datos etiquetados + muchos sin etiquetar | A aprovechar datos incompletos | Clasificar documentos cuando etiquetar es caro |
| **Aprendizaje generativo** | Grandes colecciones de ejemplos | A producir contenido nuevo parecido al aprendido | Generar texto, imágenes, audio, código |

En esta clase nos interesa entender el mapa general. En las clases siguientes vamos a trabajar sobre todo con **aprendizaje supervisado**, porque es el punto de entrada más claro para proyectos institucionales y empresariales: hay casos históricos y queremos predecir algo sobre casos nuevos.

Ejemplos de aprendizaje supervisado:

| Datos históricos | Lo que queremos predecir |
|---|---|
| Reclamos ya clasificados | Categoría de un reclamo nuevo |
| Clientes que renovaron o cancelaron | Probabilidad de abandono |
| Expedientes con tiempos de resolución | Tiempo estimado de un expediente nuevo |
| Casos de fraude confirmados | Riesgo de fraude en una operación nueva |

> Una etiqueta no es solo un dato técnico. Es una decisión humana previa: alguien definió qué categorías existen y qué respuesta era correcta.

## 5. Familias de algoritmos

Un algoritmo de ML es una forma de aprender un patrón. No hay un algoritmo “mejor” para todo: cada familia tiene fortalezas, límites y contextos donde funciona mejor.

| Familia | Idea intuitiva | Para qué suele servir | Ejemplo |
|---|---|---|---|
| **Regresiones** | Buscar una relación entre variables | Estimar valores o probabilidades | Estimar demanda o riesgo de abandono |
| **Árboles de decisión** | Hacer preguntas sucesivas tipo “si pasa esto, entonces...” | Clasificar casos con reglas interpretables | Priorizar trámites según condiciones |
| **Random Forest / Gradient Boosting** | Combinar muchos árboles para mejorar precisión | Problemas tabulares en empresas | Riesgo crediticio, fraude, churn |
| **K vecinos cercanos (KNN)** | Comparar un caso nuevo con casos parecidos | Clasificación simple por similitud | Identificar casos similares ya resueltos |
| **Naive Bayes** | Usar probabilidades de palabras o señales | Texto corto y clasificación rápida | Clasificar mensajes o tickets |
| **Clustering** | Agrupar casos parecidos sin etiquetas previas | Segmentación y exploración | Agrupar usuarios por comportamiento |
| **Redes neuronales** | Aprender representaciones complejas en muchas capas | Imágenes, audio, texto, grandes volúmenes | Detección de objetos, voz, LLMs |
| **Sistemas de recomendación** | Aprender preferencias o similitudes | Recomendar productos, contenidos o acciones | Próximo trámite sugerido, cursos recomendados |
| **Detección de anomalías** | Aprender qué es “normal” y marcar rarezas | Control, auditoría, seguridad | Transacciones inusuales, sensores fuera de rango |

Para elegir una familia no alcanza con mirar la precisión. También importan:

| Criterio | Pregunta |
|---|---|
| **Interpretabilidad** | ¿Necesito explicar por qué se tomó una decisión? |
| **Costo del error** | ¿Qué pasa si el modelo se equivoca? |
| **Cantidad de datos** | ¿Tengo suficientes ejemplos confiables? |
| **Tipo de datos** | ¿Trabajo con tablas, texto, imágenes, audio, series temporales? |
| **Velocidad** | ¿La predicción debe ocurrir en tiempo real? |
| **Mantenimiento** | ¿El patrón cambia con frecuencia? |

> En clase 8 vamos a comparar algoritmos de clasificación con evidencia. Hoy solo necesitamos reconocer que existen distintas familias y que cada una responde a una lógica diferente.

### Mini-demo: tres algoritmos, un mismo problema

Para que las familias de algoritmos no queden solo como nombres, vamos a usar un ejemplo visual y simple: clasificar una fruta como **manzana** o **naranja** según dos señales.

| Señal | Escala | Cómo leerla |
|---|---|---|
| `color_naranja` | 0 a 10 | 0 = nada naranja, 10 = muy naranja |
| `textura_rugosa` | 0 a 10 | 0 = muy lisa, 10 = muy rugosa |

La idea es intencionalmente sencilla: las manzanas del ejemplo tienden a ser menos naranjas y más lisas; las naranjas tienden a ser más naranjas y más rugosas.

Las tres celdas siguientes usan **los mismos datos**. Lo interesante es comparar cómo cambia la forma de razonar de cada algoritmo:

- KNN decide por parecido con frutas conocidas.
- El árbol de decisión aprende reglas simples.
- La regresión logística estima una probabilidad.

No hace falta modificar el código. El objetivo es observar **cómo cambia la lógica de decisión** según el algoritmo.

````python
# Algoritmo 1: KNN (vecinos cercanos)
# Idea: una fruta nueva se clasifica mirando frutas conocidas parecidas.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

frutas = pd.DataFrame([
    {"color_naranja": 1, "textura_rugosa": 1, "fruta": "manzana"},
    {"color_naranja": 2, "textura_rugosa": 2, "fruta": "manzana"},
    {"color_naranja": 3, "textura_rugosa": 1, "fruta": "manzana"},
    {"color_naranja": 3, "textura_rugosa": 3, "fruta": "manzana"},
    {"color_naranja": 4, "textura_rugosa": 2, "fruta": "manzana"},
    {"color_naranja": 2, "textura_rugosa": 3, "fruta": "manzana"},
    {"color_naranja": 7, "textura_rugosa": 7, "fruta": "naranja"},
    {"color_naranja": 8, "textura_rugosa": 8, "fruta": "naranja"},
    {"color_naranja": 9, "textura_rugosa": 8, "fruta": "naranja"},
    {"color_naranja": 8, "textura_rugosa": 9, "fruta": "naranja"},
    {"color_naranja": 10, "textura_rugosa": 7, "fruta": "naranja"},
    {"color_naranja": 7, "textura_rugosa": 9, "fruta": "naranja"},
])

frutas_nuevas = pd.DataFrame([
    {"caso": "A", "color_naranja": 2, "textura_rugosa": 2},
    {"caso": "B", "color_naranja": 8, "textura_rugosa": 8},
    {"caso": "C", "color_naranja": 5, "textura_rugosa": 5},
])

features = ["color_naranja", "textura_rugosa"]
modelo_knn = KNeighborsClassifier(n_neighbors=3)
modelo_knn.fit(frutas[features], frutas["fruta"])

resultado_knn = frutas_nuevas.copy()
resultado_knn["prediccion_knn"] = modelo_knn.predict(frutas_nuevas[features])

colores = {"manzana": "crimson", "naranja": "darkorange"}
for fruta, grupo in frutas.groupby("fruta"):
    plt.scatter(
        grupo["color_naranja"], grupo["textura_rugosa"],
        label=fruta, s=90, color=colores[fruta], alpha=0.75
    )
plt.scatter(
    frutas_nuevas["color_naranja"], frutas_nuevas["textura_rugosa"],
    marker="X", s=180, color="black", label="frutas nuevas"
)
for _, fila in frutas_nuevas.iterrows():
    plt.text(fila["color_naranja"] + 0.15, fila["textura_rugosa"], fila["caso"], fontsize=12)

plt.xlabel("Color naranja")
plt.ylabel("Textura rugosa")
plt.title("Datos conocidos y frutas nuevas")
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

print("KNN: decide por parecido con las 3 frutas conocidas más cercanas.")
display(resultado_knn)
````

````python
# Algoritmo 2: Árbol de decisión
# Idea: aprender reglas simples del tipo "si pasa esto, entonces...".

from sklearn.tree import DecisionTreeClassifier, export_text

modelo_arbol = DecisionTreeClassifier(max_depth=2, random_state=42)
modelo_arbol.fit(frutas[features], frutas["fruta"])

resultado_arbol = frutas_nuevas.copy()
resultado_arbol["prediccion_arbol"] = modelo_arbol.predict(frutas_nuevas[features])

print("Árbol de decisión: convierte los datos en reglas aproximadas.")
print()
print(export_text(modelo_arbol, feature_names=features))
display(resultado_arbol)
````

````python
# Algoritmo 3: Regresión logística
# Idea: estimar una probabilidad de pertenecer a cada categoría.

import numpy as np
from sklearn.linear_model import LogisticRegression

modelo_logistico = LogisticRegression(random_state=42)
modelo_logistico.fit(frutas[features], frutas["fruta"])

indice_naranja = list(modelo_logistico.classes_).index("naranja")
probabilidad_naranja = modelo_logistico.predict_proba(frutas_nuevas[features])[:, indice_naranja]

resultado_logistico = frutas_nuevas.copy()
resultado_logistico["probabilidad_naranja"] = np.round(probabilidad_naranja, 2)
resultado_logistico["prediccion_logistica"] = modelo_logistico.predict(frutas_nuevas[features])

print("Regresión logística: estima una probabilidad y luego decide una categoría.")
display(resultado_logistico)
````

**Para observar**

| Algoritmo | Qué mira principalmente | Qué aporta a la discusión |
|---|---|---|
| KNN | Casos parecidos cercanos | Muestra la idea de similitud |
| Árbol de decisión | Reglas simples aprendidas | Permite ver una lógica interpretable |
| Regresión logística | Probabilidad estimada | Muestra grados de confianza, no solo categorías |

Si dos algoritmos no coinciden, eso no significa que uno sea “malo” automáticamente. Significa que aprendieron aproximaciones distintas. En una aplicación real habría que medir errores, revisar casos ambiguos y decidir qué tipo de error preocupa más.

## 6. Aplicaciones en empresas e instituciones públicas

Machine Learning se usa para apoyar decisiones repetitivas, detectar patrones, priorizar trabajo y anticipar problemas. No necesariamente reemplaza personas: muchas veces funciona como una capa de ayuda para ordenar información y enfocar la atención humana.

| Área | Aplicación posible | Decisión que ayuda a tomar |
|---|---|---|
| **Atención al cliente / ciudadano** | Clasificar consultas entrantes | A qué equipo derivar cada caso |
| **Salud** | Identificar pacientes con riesgo alto | Qué casos requieren seguimiento temprano |
| **Educación** | Detectar riesgo de abandono | Qué estudiantes necesitan acompañamiento |
| **Finanzas / recaudación** | Detectar anomalías o fraude | Qué operaciones revisar |
| **Logística** | Predecir demanda o demoras | Cómo asignar stock, turnos o vehículos |
| **Recursos humanos** | Analizar rotación o clima laboral | Dónde intervenir antes de perder talento |
| **Justicia / administración pública** | Priorizar expedientes o reclamos | Qué casos tienen mayor urgencia |
| **Mantenimiento** | Predecir fallas de equipos | Cuándo hacer mantenimiento preventivo |
| **Compras y contrataciones** | Analizar proveedores y precios | Qué ofertas requieren revisión |
| **Comunicación** | Agrupar temas frecuentes en mensajes | Qué problemas comunicar o resolver primero |

Un proyecto de ML bien planteado no empieza por el algoritmo. Empieza por estas preguntas:

1. ¿Qué decisión queremos mejorar?
2. ¿Qué datos existen hoy?
3. ¿Quién generó esos datos y con qué sesgos?
4. ¿Qué error sería aceptable y cuál sería grave?
5. ¿Quién revisa los casos dudosos?
6. ¿Cómo sabremos si el sistema mejora el proceso real?

Ejemplo: si una municipalidad quiere clasificar reclamos ciudadanos, el objetivo no es “usar IA”. El objetivo puede ser reducir demoras, derivar mejor los casos y detectar temas urgentes. ML es una herramienta posible dentro de ese rediseño.

## 7. Riesgos, límites y criterios de uso

Machine Learning puede ser útil, pero no es neutral ni automático en el sentido fuerte. Aprende de datos producidos por personas, instituciones y procesos históricos. Si esos datos tienen sesgos, errores o ausencias, el modelo puede reproducirlos o amplificarlos.

| Riesgo | Qué puede pasar | Cómo mitigarlo |
|---|---|---|
| **Datos de mala calidad** | El modelo aprende patrones incorrectos | Auditar, limpiar y documentar datos |
| **Sesgos históricos** | Perjudica sistemáticamente a ciertos grupos | Medir equidad, revisar variables sensibles |
| **Falsa precisión** | Se confía demasiado en un número | Mantener revisión humana y explicar límites |
| **Cambio del contexto** | El modelo queda viejo | Monitorear resultados y reentrenar |
| **Falta de explicabilidad** | Nadie entiende por qué decide | Usar modelos interpretables cuando importa justificar |
| **Privacidad** | Se exponen datos personales o sensibles | Minimizar datos, anonimizar, controlar accesos |
| **Automatización irresponsable** | Se delegan decisiones críticas sin control | Definir supervisión, trazabilidad y mecanismos de apelación |

Una regla práctica:

| Tipo de decisión | Nivel de automatización razonable |
|---|---|
| Bajo impacto y reversible | Automatización alta puede ser aceptable |
| Impacto medio | Modelo como recomendación + revisión humana |
| Alto impacto sobre derechos, salud, empleo o acceso a servicios | Supervisión humana fuerte, explicabilidad y auditoría |

> Pregunta ética mínima: si el modelo se equivoca, ¿quién lo nota, quién responde y cómo puede corregirse el daño?

## 8. Caso guiado: clasificar consultas de usuarios

Volvamos al caso NexoDesk de las clases de prompting. El equipo recibe mensajes mezclados de usuarios y necesita ordenarlos para responder mejor.

Categorías posibles:

| Categoría | Qué significa | Ejemplo |
|---|---|---|
| **Bug** | Algo no funciona como debería | “La app se cierra cuando exporto el reporte” |
| **Sugerencia** | El usuario propone una mejora | “Sería útil tener vista de calendario” |
| **Elogio** | Comentario positivo | “La nueva versión está mucho más clara” |
| **Queja comercial** | Problema de precio, plan, cobro o atención comercial | “Me cobraron dos veces la suscripción” |

En prompting few-shot, resolvimos esto mostrando ejemplos dentro del prompt. En ML supervisado, lo pensaríamos de otra manera:

| Elemento del problema | En NexoDesk |
|---|---|
| **Casos históricos** | Mensajes recibidos anteriormente |
| **Etiqueta** | Categoría correcta asignada por una persona |
| **Patrones útiles** | Palabras, frases, tono, contexto del mensaje |
| **Predicción** | Categoría sugerida para un mensaje nuevo |
| **Uso real** | Derivar, priorizar o agrupar consultas |
| **Revisión humana** | Casos ambiguos, baja confianza o alto impacto |

Ejemplo de dataset conceptual:

| Mensaje | Etiqueta humana |
|---|---|
| “Cada vez que intento exportar el reporte en PDF, la app se cierra.” | Bug |
| “Podrían agregar recordatorios por WhatsApp para tareas vencidas.” | Sugerencia |
| “La última versión está mucho más rápida, gracias.” | Elogio |
| “El plan Pro cuesta el doble que la competencia.” | Queja comercial |

Lo importante no es el código. Lo importante es reconocer el proceso de pensamiento:

1. Definir qué decisión queremos apoyar: derivar mensajes.
2. Definir categorías claras.
3. Reunir ejemplos reales.
4. Etiquetar con criterio consistente.
5. Entrenar un sistema que aprenda patrones.
6. Medir errores.
7. Revisar casos dudosos.
8. Mejorar datos y criterios.

### Comparación con prompting

| Pregunta | Prompting few-shot | ML supervisado |
|---|---|---|
| ¿Dónde están los ejemplos? | En el prompt | En un dataset |
| ¿Quién define el criterio? | La instrucción y los ejemplos elegidos | Las etiquetas históricas |
| ¿Cómo se mejora? | Iterando el prompt | Mejorando datos, etiquetas, modelo y evaluación |
| ¿Qué pasa con casos ambiguos? | Pedimos aclaraciones o reglas | Marcamos baja confianza o revisión humana |
| ¿Cuándo conviene? | Pocos casos, flexibilidad, lenguaje abierto | Muchas repeticiones, tarea estable, medición sistemática |

### Preguntas para discutir

1. ¿Qué categorías están bien definidas y cuáles podrían confundirse?
2. ¿Qué mensaje podría ser a la vez `Sugerencia` y `Queja comercial`?
3. ¿Qué errores serían más graves para el negocio o la institución?
4. ¿Cuántos ejemplos reales necesitaríamos antes de confiar en el sistema?
5. ¿Qué casos deberían ir siempre a revisión humana?
