# Clase 10 — Presentación y defensa de decisiones en ML

En esta clase cerramos el recorrido iniciado en la clase 6. El foco no está en programar, sino en **presentar, argumentar y defender decisiones** sobre problemas de Machine Learning usando el vocabulario y los criterios trabajados en el módulo.

Cada grupo retoma uno de los casos analizados en la clase 9, explica qué incoherencia detectó, propone una corrección y comparte de qué manera usó la inteligencia artificial para comprender mejor el problema sin delegar la decisión final.

## 1. Qué aprendimos desde la clase 6 hasta la 9

A lo largo de estas clases construimos un recorrido progresivo:

- **Clase 6:** qué es Machine Learning, qué problemas resuelve y qué familias de algoritmos existen.
- **Clase 7:** cuál es el flujo estándar de un proyecto de ML: problema → dataset → split → preprocesamiento → entrenamiento → evaluación.
- **Clase 8:** cómo comparar modelos de clasificación y elegir uno con evidencia.
- **Clase 9:** cómo detectar errores conceptuales en decisiones de ML y corregirlos con apoyo de IA.

Hoy el objetivo es integrar todo eso en una presentación clara, breve y argumentada.

---
## 2. Objetivo de la presentación grupal

Cada grupo presenta un caso trabajado en la clase 9. La meta no es mostrar una respuesta perfecta, sino demostrar que pueden **leer una situación**, **reconocer una incoherencia** y **justificar una corrección** con conceptos del módulo.

La exposición debe dejar claro:
1. qué problema se quería resolver;
2. cuál era la decisión incorrecta o incompleta;
3. cómo la corregirían;
4. qué concepto del módulo respalda esa corrección;
5. cómo usaron inteligencia artificial para entender mejor el caso.

---
## 3. Guía para defender una corrección

Cuando expliquen su caso, conviene seguir este orden:

1. **Definir el problema**
   - ¿Es clasificación o regresión?
   - ¿Qué se quiere predecir o decidir?

2. **Marcar la incoherencia**
   - ¿El algoritmo elegido no correspondía?
   - ¿La métrica no tenía sentido en ese contexto?
   - ¿Se ignoró una señal de sobreajuste?
   - ¿Faltó una etapa del flujo estándar?

3. **Proponer una corrección**
   - ¿Qué cambiarían?
   - ¿Qué alternativa sería más razonable?

4. **Justificar**
   - ¿Qué idea de las clases 6 a 9 respalda la corrección?
   - ¿Qué tipo de error sería más costoso en ese problema?

---
## 4. Cómo usar IA como apoyo crítico

La inteligencia artificial puede ayudar mucho si se usa como herramienta para pensar mejor. Por ejemplo, puede servir para:

- pedir explicaciones más simples de conceptos técnicos;
- comparar dos métricas o dos algoritmos;
- revisar si una justificación está clara;
- sugerir una mejor forma de redactar la conclusión del grupo.

No debería usarse para delegar toda la respuesta. La parte importante del ejercicio sigue siendo el **criterio del grupo**.

### Ejemplos de prompts útiles

- "Explicame como si recién empezara la diferencia entre accuracy y recall"
- "¿Por qué este caso parece de regresión y no de clasificación?"
- "Ayudame a revisar si esta justificación tiene sentido técnico"
- "Reescribí esta explicación para que sea más clara, pero sin cambiar la idea"

Completen esta guía antes de exponer:

### Sugerencia de tiempo

- 2 minutos para explicar el caso
- 2 minutos para mostrar el error detectado
- 1 minuto para defender la corrección propuesta

## 6. Cierre y reflexión final

Al terminar las presentaciones, cada grupo puede cerrar con una reflexión breve:

1. ¿Qué fue lo más difícil del caso: reconocer el problema, elegir la métrica o justificar la corrección?
2. ¿Qué concepto del módulo les quedó más claro después de discutirlo en grupo?
3. ¿En qué momento la IA fue útil y en cuál no alcanzaba por sí sola?

> Objetivo final del módulo: no memorizar nombres de algoritmos, sino empezar a desarrollar criterio para elegir, cuestionar y justificar decisiones en contextos reales.
