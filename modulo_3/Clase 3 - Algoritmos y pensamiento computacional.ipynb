## 1. ¿Qué es un algoritmo?

Un **algoritmo** es una secuencia finita de pasos bien definidos para resolver un problema.

No tiene que ser matemático. Un algoritmo puede servir para:

- Preparar un desayuno según el tiempo disponible.
- Decidir si hay que regar una planta.
- Elegir cómo sacar una foto.
- Clasificar una fruta.
- Decidir si una consulta está lista para enviarse a un modelo local.

La idea central es siempre la misma: transformar una situación ambigua en pasos claros.

| Parte | Pregunta que responde | Ejemplo |
|---|---|---|
| Entrada | ¿Qué datos necesito? | `tiempo_disponible`, `ingredientes`, `preferencia` |
| Proceso | ¿Qué hago con esos datos? | Comparar, ordenar, validar, elegir |
| Decisión | ¿Qué condición cambia el camino? | `tiempo_disponible >= 10` |
| Salida | ¿Qué resultado entrego? | `preparar tostadas` |


---

## 2. Propiedades de un buen algoritmo

| Propiedad | Qué significa | Pregunta de control |
|---|---|---|
| **Finito** | Tiene inicio y fin claros | ¿Puede terminar siempre? |
| **Determinista** | Con los mismos datos produce el mismo resultado | ¿Dos personas llegarían a la misma salida? |
| **General** | Sirve para muchos casos, no solo para uno | ¿Funciona si cambian los datos? |
| **Completo** | Cubre los casos esperables | ¿Qué pasa si ninguna condición se cumple? |

Un algoritmo incompleto suele fallar por una de estas razones:

- Le falta una entrada importante.
- Tiene una condición ambigua.
- No define qué hacer en un caso límite.
- No tiene una salida clara.

## 5. Del algoritmo al modelo local

En la Clase 10 vas a trabajar con un modelo de chat local. Aunque parezca un tema distinto, la lógica de control es la misma.

Antes de llamar a un modelo necesitamos decidir cosas como:

| Decisión | Entrada necesaria | Salida posible |
|---|---|---|
| ¿El prompt está listo? | texto del usuario, longitud, claridad | enviar / pedir aclaración |
| ¿Qué rol usar? | tipo de tarea | `system`, `user` |
| ¿Qué historial enviar? | mensajes anteriores | incluir historial / empezar nuevo chat |
| ¿Qué temperatura usar? | creatividad esperada | baja / media / alta |
| ¿Conviene streaming? | longitud esperada de respuesta | usar streaming / esperar respuesta completa |
| ¿Hay que revisar la respuesta? | sensibilidad del tema, confianza | revisar / aceptar |

```python
# Ejecuta esta celda una vez antes de resolver las actividades.
import unicodedata

SOLUCIONES_ESPERADAS = {
    "actividad_1": [
        "comer fruta",
        "preparar tostadas",
        "comprar algo afuera",
        "hacer cafe",
    ],
    "actividad_2": [
        "regar ahora",
        "no regar",
        "no regar",
        "esperar",
    ],
    "actividad_3": [
        "usar flash",
        "usar modo movimiento",
        "acercarse",
        "sacar foto normal",
    ],
    "actividad_4": [
        "manzana",
        "naranja",
        "desconocida",
    ],
    "actividad_5": [
        "pedir aclaracion",
        "cargar modelo primero",
        "enviar al modelo",
        "acortar prompt",
    ],
}


def normalizar_respuesta(texto):
    if not isinstance(texto, str):
        return ""
    texto = unicodedata.normalize("NFKD", texto.strip().lower())
    texto = "".join(ch for ch in texto if not unicodedata.combining(ch))
    return " ".join(texto.split())



def revisar_actividad(clave, nombre, pasos, respuestas_casos, casos):
    esperadas = SOLUCIONES_ESPERADAS[clave]

    print(f"=== {nombre} ===")
    print()
    print("Algoritmo en texto:")
    faltan_pasos = False
    for i, paso in enumerate(pasos, start=1):
        if isinstance(paso, str) and paso.strip():
            print(f"Paso {i}: {paso}")
        else:
            faltan_pasos = True
            print(f"Paso {i}: [completar]")

    print()
    print("Chequeo de casos de prueba:")
    faltan_respuestas = False
    if len(respuestas_casos) != len(esperadas):
        print(
            f"Se esperaban {len(esperadas)} respuestas y se recibieron {len(respuestas_casos)}."
        )
        faltan_respuestas = True

    for i, (caso, esperada) in enumerate(zip(casos, esperadas), start=1):
        if i - 1 < len(respuestas_casos):
            respuesta = respuestas_casos[i - 1]
        else:
            respuesta = ""

        print(f"Caso {i}: {caso}")
        if isinstance(respuesta, str) and respuesta.strip():
            coincide = normalizar_respuesta(respuesta) == normalizar_respuesta(esperada)
            estado = "OK" if coincide else f"Revisar -> esperaba {esperada}"
            print(f"  respuesta: {respuesta}")
            print(f"  resultado: {estado}")
        else:
            faltan_respuestas = True
            print("  respuesta: [completar]")
            print("  resultado: pendiente")

    if faltan_pasos or faltan_respuestas:
        print()
        print("Completa los campos pendientes y vuelve a ejecutar la celda.")
    else:
        print()
        print("Actividad completa.")


print("Celda auxiliar lista. Ahora podes completar las actividades.")
```
