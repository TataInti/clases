## 1. ¿Por qué NumPy? De listas a arrays

Con una lista, para sumar 1 grado a cada temperatura de la semana hay que recorrerla elemento por elemento:

```python
temperaturas = [18, 21, 19, 23, 20, 17, 22]
mas_un_grado = []
for t in temperaturas:
    mas_un_grado.append(t + 1)
```

Con un **array de NumPy**, la misma operación se escribe directamente sobre todos los valores a la vez:

```python
temperaturas = np.array([18, 21, 19, 23, 20, 17, 22])
mas_un_grado = temperaturas + 1
```

Esto se llama **vectorización**: en vez de decirle a Python "repetí esto para cada elemento", operamos el array completo de una sola vez. Es más corto de escribir y, con datasets grandes, mucho más rápido.


````python
import numpy as np

temperaturas_lista = [18, 21, 19, 23, 20, 17, 22]

# Con lista: hay que recorrer elemento por elemento
mas_un_grado_lista = []
for t in temperaturas_lista:
    mas_un_grado_lista.append(t + 1)

# Con array: la operación se aplica a todos los valores de una vez
temperaturas_array = np.array(temperaturas_lista)
mas_un_grado_array = temperaturas_array + 1

print("Con lista:", mas_un_grado_lista)
print("Con array:", mas_un_grado_array)
print("Tipo del array:", type(temperaturas_array))
````

## 2. Crear arrays

Hay varias formas de crear un array, según lo que necesitemos:

| Función | Qué hace |
|---|---|
| `np.array([...])` | Convierte una lista (o lista de listas) en array |
| `np.zeros(n)` | Crea un array de `n` ceros |
| `np.ones(n)` | Crea un array de `n` unos |
| `np.arange(inicio, fin, paso)` | Crea una secuencia, igual que `range()` pero como array |
| `np.linspace(inicio, fin, cantidad)` | Crea `cantidad` valores repartidos parejo entre `inicio` y `fin` |


````python
desde_lista = np.array([10, 20, 30, 40])
ceros = np.zeros(5)
unos = np.ones(3)
secuencia = np.arange(0, 10, 2)
repartidos = np.linspace(0, 1, 5)

print("Desde lista :", desde_lista)
print("Ceros       :", ceros)
print("Unos        :", unos)
print("Secuencia   :", secuencia)
print("Repartidos  :", repartidos)
````

## 3. Atributos de un array

Todo array de NumPy sabe describirse a sí mismo:

| Atributo | Qué informa |
|---|---|
| `shape` | Forma del array: filas y columnas |
| `dtype` | Tipo de dato guardado (`int64`, `float64`, etc.) |
| `ndim` | Cantidad de dimensiones (1D, 2D, ...) |
| `size` | Cantidad total de elementos |


````python
notas = np.array([7.5, 8.0, 6.5, 9.0, 5.5])

print("Array   :", notas)
print("shape   :", notas.shape)
print("dtype   :", notas.dtype)
print("ndim    :", notas.ndim)
print("size    :", notas.size)
````

## 4. Indexing y slicing

Para arrays de **1 dimensión**, indexar y recortar funciona igual que con listas (Clase 7): posición desde 0, y `[inicio:fin]` para un rango.


````python
notas = np.array([7.5, 8.0, 6.5, 9.0, 5.5])

print("Primera nota      :", notas[0])
print("Última nota       :", notas[-1])
print("Primeras 3 notas  :", notas[0:3])
print("Notas desde la 2  :", notas[2:])
````

Cuando el array tiene **2 dimensiones** (filas y columnas), se indexa con `[fila, columna]`. Pensemos en una matriz de notas: cada fila es un alumno, cada columna es una materia.


````python
# Filas = alumnos, columnas = materias (Matemática, Lengua, Ciencias)
notas_curso = np.array([
    [7.5, 8.0, 6.5],
    [9.0, 5.5, 7.0],
    [6.0, 6.5, 8.5],
])

print("Forma de la matriz          :", notas_curso.shape)
print("Nota del alumno 0 en Lengua :", notas_curso[0, 1])
print("Todas las notas del alumno 1:", notas_curso[1])
print("Todas las notas de Ciencias :", notas_curso[:, 2])
````

## 5. Operaciones vectorizadas y broadcasting

Ya vimos que se puede sumar un número a todo un array (`temperaturas + 1`). Eso se llama **broadcasting**: NumPy "reparte" el número entre todos los elementos automáticamente. Lo mismo funciona con `-`, `*`, `/` y con comparaciones (`>`, `<`, `==`).


````python
notas = np.array([7.5, 8.0, 6.5, 9.0, 5.5])

notas_sobre_100 = notas * 10          # escalar multiplicando a todo el array
aprobo = notas >= 7                    # comparación → array de booleanos

print("Notas x10   :", notas_sobre_100)
print("¿Aprobó?    :", aprobo)
print("Solo aprobados:", notas[aprobo])   # indexar con un array de booleanos
````

## 6. Funciones estadísticas

En vez de escribir un loop para sumar o promediar, NumPy trae funciones listas:

| Función | Qué calcula |
|---|---|
| `.sum()` | Suma de todos los valores |
| `.mean()` | Promedio |
| `.min()` / `.max()` | Valor mínimo / máximo |
| `.std()` | Desviación estándar (qué tan dispersos están los valores) |


````python
temperaturas = np.array([18, 21, 19, 23, 20, 17, 22])

print("Suma    :", temperaturas.sum())
print("Promedio:", temperaturas.mean())
print("Mínima  :", temperaturas.min())
print("Máxima  :", temperaturas.max())
print("Desvío  :", round(temperaturas.std(), 2))
````
