# Clase 6: Funciones en Python
En esta clase aprenderemos qué son las funciones, por qué son importantes y cómo nos ayudan a escribir código más limpio, reutilizable y organizado.

## ¿Qué es una función en programación?
Una función es un **bloque de código con nombre propio** que se puede ejecutar (llamar) cuando lo necesitemos. Nos permite **organizar, reutilizar y simplificar** nuestros programas.

### Veamos algunos ejemplos

````python
def saludar():
    print("Hola, bienvenido a Python")
````

````python
saludar()
````

###¿Qué elementos forman una función en Python?

Una función en Python está compuesta por varios componentes esenciales.

*Explicaciones uno por uno:*

1. **Definición con `def`**

  Es la palabra clave que le indica a Python que estamos definiendo una función.

  `def saludar():`

2. **Nombre de la función**

  Es el identificador que se usa para llamar a esa función más adelante.
  `def mostrar_mensaje():`
    ...

  📌 Regla: debe seguir las reglas de nombres de variables (sin espacios, sin empezar con número, etc.).

3. **Paréntesis y parámetros** (opcionales)

 Dentro de los paréntesis se colocan los parámetros, es  decir, los datos que la función necesita para trabajar.  
 Pueden ser uno, varios o ninguno.
  ```
  def saludar(nombre):
    print("Hola", nombre)
    ```
 🔸 Si no tiene parámetros:
```
 def bienvenida():
    print("¡Bienvenido!")
    ```
🔸 Si tiene varios:
 ```
 def sumar(a, b):
     return a + b
```
4. **Dos puntos `:`**

 Al final de la línea de definición, siempre va un : para indicar que comienza el bloque de código de la función.

5. **Cuerpo de la función** (el código indentado)

 Todo el código que está indentado (sangria hacia la derecha) forma parte del cuerpo de la función. Es lo que se ejecuta cuando llamamos la función.
```
 def mostrar_nombre():
    nombre = "Ana"
    print("Mi nombre es", nombre)
```

````python
# ingresamos las variables que usaremos en la función
nombre = input("ingrese su nombre: ")
edad = input("ingrese su edad: ")
````

````python
# creamos la funcion que va a recoger el valor de 2 variables y
# las va a imprimir con un mensaje

def saludo(x, y):
  print("Hola soy", x,"y tengo", y, "años")
````

````python
# llamamos la funcion y le pasamos los argumentos para que funcione
saludo(nombre, edad)
````

````python
# Hay que ser cuidadoso con el orden de los argumentos
saludo(edad, nombre)
````

````python
# la funcion espera un minimo de argumentos para ejecutarse
saludo(nombre)
````

### 🔹 Funciones con parametros

````python
def saludar(nombre_user):
    print("Hola", nombre_user)
````

````python
saludar("Ana")  # Output: Hola Ana
````

### 🔹 Funciones que devuelven un valor `(return)`

````python
# definir la funcion
def sumar(a, b):
    return a + b
````

````python
i = int(input("ingrese un numero "))
j = int(input("ingrese un numero "))

resultado = sumar(i, j)
````

````python
print("La suma es:", resultado)
````

## ¿Qué puede tener una función?
- **Nombre**: cómo la vamos a llamar.
- **Parámetros**: datos que recibe (opcionales).
- **Código interno**: instrucciones que ejecuta.
- **Valor de retorno**: resultado que devuelve (opcional).

## Actividad 2: Calcular el cuadrado
Escribir una función `cuadrado(x)` que devuelva el cuadrado de un número.

````python
# definir la logica de la función
def obterner_valor_al_cuadrado(parametro1):
  return parametro1 * parametro1


# valores para ser calculados
variable1 = int(input("ingrese un numero: "))

````

````python
# usar la funcion
print("El valor cuadrado de", variable1, "es", obterner_valor_al_cuadrado(variable1))
````

````python
# primero definimos la funcion
def cuadrado(x):
    return x * x

# generamos los argumentos
valor1 = int(input("Ingresa un numero"))

# usamos la funcion
print(f"El cuadrado de {valor1} es:", cuadrado(valor1))
````

## Actividad 3: Verificar si un número es par
Crear una función `es_par(n)` que devuelva `True` si el número es par y `False` si no lo es.

````python
# creamos la funcion como primer paso
def es_par(x):
    return x % 2 == 0    # la op matemática divide el num ingresado y busca que el resto sea igual a 0

# obtenemos los argumentos
i = int(input("Ingresa un numero"))

# usamos la funcion
es_par(i)
# print(f"¿{i} es par?:", es_par(i))
````
