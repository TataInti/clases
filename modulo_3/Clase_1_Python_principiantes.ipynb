## ¿Qué es la Programación?

La programación es el proceso de diseñar y escribir instrucciones que una computadora puede ejecutar para realizar una tarea específica. Piensa en ello como la creación de un conjunto de órdenes lógicas, similares a un algoritmo o un procedimiento que resuelve un problema. Por ejemplo, al seguir una receta para preparar un plato, estás proporcionando pasos secuenciales; de manera similar, al programar, defines pasos que la computadora debe seguir.

En términos más técnicos, la programación implica traducir una idea o solución a un lenguaje que la máquina entienda. Estos lenguajes de programación, como Python, actúan como intermediarios entre el pensamiento humano y el hardware de la computadora.

**Relevancia de la programación:**  
- Es una habilidad transversal que se aplica en ciencias, ingeniería, economía, artes digitales y más.  
- Permite automatizar tareas repetitivas, analizar datos, construir aplicaciones y resolver problemas complejos.

## Tipos de Datos en Python

Ahora que entendemos las variables, es importante conocer los diferentes **tipos de datos** que podemos almacenar en ellas. Python es un lenguaje dinámicamente tipado, lo que significa que no necesitas declarar el tipo de dato de una variable explícitamente; Python lo infiere según el valor asignado. Sin embargo, conocer los tipos de datos es crucial para manipular la información correctamente.

Vamos a explorar los tipos de datos fundamentales de Python, paso a paso, con ejemplos que ilustran su uso y propósito:

1. **Cadenas de Texto (str - string):** Representan texto o caracteres. Se escriben entre comillas simples (`'`) o dobles (`"`). Ejemplo: `"Hola"` o `'Python'`. Son útiles para manejar nombres, mensajes o cualquier dato textual.
2. **Números Enteros (int - integer):** Representan números enteros, positivos o negativos, sin parte decimal. Ejemplo: `5`, `-10`. Se usan para conteos o valores discretos.
3. **Números Decimales (float - floating point):** Representan números con decimales. Ejemplo: `3.14`, `0.5`. Son comunes en cálculos matemáticos o mediciones.
4. **Booleanos (bool):** Representan valores de verdad, es decir, `True` (verdadero) o `False` (falso). Son fundamentales para la lógica de decisión en programas (los usaremos más adelante con condicionales).
5. **Listas (list):** Colecciones ordenadas y modificables de elementos, que pueden ser de diferentes tipos. Ejemplo: `["manzana", "banana", 3]`. Son útiles para manejar conjuntos de datos.
6. **Diccionarios (dict):** Colecciones de pares clave-valor, ideales para organizar datos estructurados. Ejemplo: `{"nombre": "Ana", "edad": 20}`. Se usan para representar información relacionada, como registros.
7. **Ninguno (NoneType):** Representa la ausencia de valor, como un contenedor vacío. Se escribe como `None`. Es útil para inicializar variables o indicar que no hay dato disponible.

**Nota:** En esta clase, nos enfocaremos principalmente en cadenas de texto, números y booleanos. Los tipos como listas y diccionarios los exploraremos con más detalle en sesiones futuras. Por ahora, el objetivo es que tengas una visión general de las posibilidades que Python ofrece.

En la celda siguiente, ejecuta el código para ver ejemplos de cada tipo de dato y cómo se muestran en la salida.

````python
print(usuario)
````

````python
# Ejemplos de los diferentes tipos de datos en Python
texto = "¡Hola a todos, soy una cadena de texto!"  # String (texto)
numero_entero = 42  # Integer (número entero)
numero_decimal = 3.14  # Float (número decimal)
es_verdad = True  # Boolean (verdadero/falso)
lista_compras = ["manzana", "banana", "leche", 3]  # List (lista)
info_persona = {"nombre": "Ana", "edad": 20}  # Dictionary (diccionario)
nada = None  # None (sin valor)

# Mostramos cada tipo de dato con print()
print("Cadena de texto (str):", texto)
print("Número entero (int):", numero_entero)
print("Número decimal (float):", numero_decimal)
print("Booleano (bool):", es_verdad)
print("Lista (list):", lista_compras)
print("Diccionario (dict):", info_persona)
print("Valor None (NoneType):", nada)

# Podemos verificar el tipo de dato usando la función type()
print("\nVerificando tipos de datos con type():")
print("Tipo de texto:", type(texto))
print("Tipo de numero_entero:", type(numero_entero))
````

## Análisis de los Tipos de Datos

En el ejemplo anterior, hemos creado variables con diferentes tipos de datos y las hemos mostrado en la consola. Además, usamos la función `type()` para inspeccionar el tipo de dato de cada variable, lo cual es útil para depurar y entender cómo Python interpreta nuestros valores.

**Observaciones clave:**  
- Las cadenas de texto siempre están entre comillas. Si olvidas las comillas, Python intentará interpretar el texto como una variable o comando, lo que resultará en un error.  
- Los números (`int` y `float`) no llevan comillas. La diferencia entre entero y decimal es que el `float` incluye un punto decimal, incluso si es algo como `5.0`.  
- Los booleanos (`True` y `False`) son palabras reservadas en Python y deben escribirse exactamente así, con la primera letra en mayúscula. Representan estados binarios y son fundamentales para la lógica de control en programas.  
- Las listas y diccionarios son estructuras de datos más complejas. Una lista es como un arreglo ordenado donde puedes almacenar múltiples valores, mientras que un diccionario organiza datos en pares clave-valor, similar a una base de datos pequeña.  
- `None` indica "ausencia de valor". Es útil en situaciones donde una variable no tiene un dato asignado o para inicializar algo que se definirá más adelante.

**Importancia de los tipos de datos:**  
El tipo de dato determina qué operaciones puedes realizar con una variable. Por ejemplo, puedes sumar dos números (`int` o `float`), pero no puedes sumar un número y una cadena de texto directamente. Conocer los tipos de datos te ayudará a evitar errores y a escribir código más eficiente.
