# Clase 7 — Listas, diccionarios y datasets para IA

En la Clase 6 usamos `for` para recorrer datos muchas veces. Pero para recorrer datos primero necesitamos **guardarlos** de una forma ordenada.

Hoy vamos a trabajar con dos estructuras muy usadas en Python:
- **listas**, para guardar varios elementos en orden,
- **diccionarios**, para guardar un registro con nombre para cada dato.

Después vamos a unir ambas ideas para representar un dataset como una **lista de diccionarios**.

---

## 1. Listas: guardar varios datos en orden

Una lista sirve para guardar varios valores dentro de una sola variable.

Ejemplo:
- una lista de pesos,
- una lista de nombres,
- una lista de frutas observadas.

En una lista, cada elemento tiene una **posición**. Esa posición se llama **índice**.

```python
# Una lista simple
frutas = ["manzana", "banana", "pera"]


print(frutas)
print("Primera fruta:", frutas[0])
print("Segunda fruta:", frutas[1])
print("Cantidad total:", len(frutas))
```

---

## 2. Modificar una lista

Las listas no son fijas. Podemos:
- cambiar un elemento,
- agregar un elemento nuevo,
- recorrer toda la lista con un `for`, como vimos en la Clase 6.

Usamos el metodo `append()` para agregar un elemento al final de la lista.


```python
# Cambiar y agregar elementos en una lista
frutas = ["manzana", "banana", "pera"]

frutas[2] = "naranja"
frutas.append("uva")  

print("Lista actualizada:", frutas)

for fruta in frutas:
    print(fruta)
```

```python
lista = [0, 1, [2, 3, 4]]
lista[[0][0]] = 1

print(lista)
```

---

## 3. Diccionarios: un dato con varias características

Una lista sirve para guardar varios elementos. Pero a veces un solo elemento tiene varias partes.

Por ejemplo, una fruta puede tener:
- `nombre`
- `color`
- `peso`

Para eso usamos un **diccionario**. En lugar de acceder por posición, accedemos por **clave**.

```python
# Un diccionario representa un registro
fruta = {"nombre": "uva", "color": "morado"}

print(fruta)
print("Nombre:", fruta["nombre"])
print("Color:", fruta["color"])
print("Peso:", fruta["peso"])
```

---

## 4. Leer sin romper: `.get()`

Si intentamos leer una clave que no existe con corchetes, Python da error.

Cuando no estamos seguros de que el dato exista, usamos `.get()`.
Si la clave falta, podemos devolver un valor por defecto.

```python
# Lectura segura con .get()
fruta = {"nombre": "pera", "color": "verde"}

print("Nombre:", fruta.get("nombre"))
print("Peso:", fruta.get("peso", 0))
print("Tipo:", fruta.get("tipo", "desconocido"))

print("Diccionario completo:", fruta)
```

---

## 5. Modificar un diccionario

También podemos cambiar datos existentes o agregar claves nuevas.

Esto se parece a lo que hicimos con variables en clases anteriores: leer, decidir y actualizar.

```python
# Cambiar y agregar datos en un diccionario
fruta = {"nombre": "banana", "color": "amarillo"}

fruta["color"] = "amarillo claro"
fruta["peso"] = 180

print("Registro actualizado:", fruta)
```

```python
# Cambiar y agregar datos en un diccionario
fruta2 = [{"nombre": "banana", "color": "amarillo"},
         {"nombre": "banana", "color": "amarillo"}]

fruta2[0]["color"] = "amarillo claro"


print("Registro actualizado:", fruta2)
```

---

## 6. Un dataset: una lista de diccionarios

Ahora unimos las dos ideas.

- Cada **diccionario** representa una fruta.
- La **lista** contiene todas las frutas.

Eso es exactamente un dataset pequeño en memoria.

```python
# Dataset pequeño: lista de diccionarios
dataset = [
    {"nombre": "manzana", "color": "rojo", "peso": 150},
    {"nombre": "banana", "color": "amarillo", "peso": 180},
    {"nombre": "pera", "color": "verde", "peso": 120}
]

print("Cantidad de registros:", len(dataset))
print("Primer registro:", dataset[2])
```

---

## 7. Recorrer un dataset con `for`

Acá se conecta directamente con la Clase 6.

Ya sabemos recorrer una lista. Como el dataset es una lista, podemos usar `for` para mirar cada registro uno por uno.

```python
# Recorrer el dataset
dataset = [
    {"nombre": "manzana", "color": "rojo", "peso": 150},
    {"nombre": "banana", "color": "amarillo", "peso": 180},
    {"nombre": "pera", "color": "verde", "peso": 120}
]

for dato in dataset:
    print(dato["nombre"], "pesa", dato["peso"], "gramos")
```

---

## 8. Detectar datos faltantes

En la práctica, no todos los registros vienen completos.

Por eso conviene revisar si falta información importante, por ejemplo el `peso`.

```python
# Detectar faltantes con .get()
dataset = [
    {"nombre": "manzana", "color": "rojo", "peso": 150},
    {"nombre": "banana", "color": "amarillo"},
    {"nombre": "pera", "color": "verde", "peso": 120}
]

for dato in dataset:
    peso = dato.get("peso", 0)
    
    if peso == 0:
        print("Falta el peso de:", dato["nombre"])
    else:
        print(dato["nombre"], "tiene peso registrado")
```

## 9. Mini ETL muy simple

Vamos a hacer una limpieza mínima:
- leer cada registro,
- copiarlo,
- completar el `peso` con `0` si falta,
- guardar el resultado en una nueva lista.

Esto es una versión muy sencilla de una ETL en memoria.

```python
# Mini ETL simple
dataset_sucio = [
    {"nombre": "manzana", "color": "rojo", "peso": 150},
    {"nombre": "banana", "color": "amarillo"},
    {"nombre": "pera", "color": "verde", "peso": 120}
]

dataset_limpio = []

for registro in dataset_sucio:
    nuevo_registro = registro.copy()
    
    if "peso" not in nuevo_registro:
        nuevo_registro["peso"] = 0
    
    dataset_limpio.append(nuevo_registro)

print("Original:", dataset_sucio)
print("Limpio:", dataset_limpio)
```
