## ¿Qué es una lista?
Una lista es una colección ordenada y modificable de elementos. Puede contener datos de cualquier tipo (números, cadenas, otras listas…).

 Se define con corchetes `[]` y puede contener distintos tipos de datos.

````python
# Creamos una variable que contiene una lista
tareas = ["Estudiar", "Comprar pan", "Llamar al médico", input("Agrega una tarea: ")]
print(tareas)
````

````python
# Ahora podemos recorrerla y extraer los datos
for tarea in tareas:
    print("Pendientes:", tarea)
````

## Listas: cuando el orden importa
Una lista es como una fila de objetos en una estantería: podemos acceder a cada elemento por su posición (índice), recorrerlos, agregar o quitar cosas, etc.

Ejemplos de uso real:
- Lista de productos en un carrito de compras.
- Lista de respuestas de un cuestionario.
- Lista de tareas pendientes.
Nombres de alumnos en una clase.

````python
# Crear una lista de ejemplo
frutas = ["manzana", "banana", "naranja"]
print(frutas[2])
````

## Modificar listas
- `.append()` agrega un elemento
- `.remove()` elimina un elemento
- `.insert()` agrega en una posición específica

````python
#Agregar un dato
frutas.append("kiwi")
print(frutas)
````

````python
# Eliminar un dato
frutas.remove("banana")
print(frutas)
````

## Recorrer una lista con `for`

````python
for fruta in frutas:
    print("En esta lista tenemos:", fruta)
````

## ¿Qué es un diccionario?
Es una colección **no ordenada** de pares `clave: valor`.


*   Se define con llaves `{}`.
*   Se accede mediante las etiquetas o claves.



````python
# Crear un diccionario
persona = {"nombre": input("Ingresa tu nombre"), "edad": 20, "carrera": "Ingeniería"}
print("Nombre:", persona["nombre"])
print("Carrera:", persona["carrera"])
````

````python
persona = {
    "nombre": input("Ingresa tu nombre"),
    "edad": 20,
    "carrera": "Ingeniería",
    "materias": ["matematicas", "historia"]
    }

print("Nombre:", persona["nombre"])
print("Carrera:", persona["carrera"])
print("Materias:", persona["materias"])
````

## Agregar, modificar o eliminar elementos en diccionarios

````python
# Agregar un par clave:valor
persona["ciudad"] = "Tucumán"
print(persona)
````

````python
# Modificar un registro
persona["edad"] = 21
print(persona)
````

````python
#Eliminar un par clave:valor
del persona["carrera"]
print(persona)
````

## Recorrer un diccionario

````python
persona = {
    "nombre": input("Ingresa tu nombre"),
    "edad": 20,
    "carrera": "Ingeniería",
    "materias": ["matematicas", "historia"]
    }
````

````python
for cada_registro in persona:
    print(cada_registro, "=", persona[cada_registro])
````

## Combinación poderosa: `listas de diccionarios`

Cuando tenemos muchos objetos estructurados, usamos una lista de diccionarios.

Por ejemplo, una agenda de contactos, un catálogo de productos, o una base de datos simple.

```
agenda = [
    {"nombre": "Ana", "telefono": "123456"},
    {"nombre": "Luis", "telefono": "654321"},
    {"nombre": "Sofía", "telefono": "112233"}
]

for persona in agenda:
    print(persona["nombre"], "->", persona["telefono"])
```

````python
agenda = [
    {"nombre": "Ana", "telefono": "123456"},
    {"nombre": "Luis", "telefono": "654321"},
    {"nombre": "Sofía", "telefono": "112233"}
]

for persona in agenda:
    print(persona["nombre"], "->", persona["telefono"])
````
