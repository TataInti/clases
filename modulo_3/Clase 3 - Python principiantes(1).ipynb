# Clase 3: Decisiones y Condicionales en Python
En esta clase aprenderemos a tomar decisiones dentro de un programa utilizando estructuras condicionales como `if`, `else` y `elif`.
Esto nos permitirá crear programas que reaccionan de forma distinta según los datos.

## ¿Qué es un Algoritmo?

Un algoritmo es una secuencia ordenada de pasos lógicos que se siguen para resolver un problema o realizar una tarea. Es como una receta de cocina:
- Tiene un inicio y un fin.
- Cada paso debe ser claro y ejecutable.
- Está diseñado para alcanzar un objetivo específico.

**Ejemplo cotidiano:**

_"Quiero decidir qué ropa usar hoy"_

Algoritmo:
1. Mirar el clima.
2. Si está soleado, usar ropa ligera.
3. Si está nublado o llueve, usar abrigo y paraguas.
4. Salir de casa.

### Algoritmos en la Programación
Cuando escribimos código en Python, estamos traduciendo esos pasos lógicos a instrucciones que una computadora puede ejecutar.

Por ejemplo, en esta clase aprendemos a usar condicionales (`if`, `else`, `elif`), que son la forma que tiene un algoritmo para tomar decisiones:

````python
temperatura = 25

if temperatura > 25:
    print("Usa ropa ligera")
else:
    print("Lleva abrigo")
````

Este código es la versión programada de un pequeño algoritmo que responde a la pregunta: ¿Qué ropa usar según la temperatura?

## ¿Qué es una estructura condicional?
Una estructura condicional permite ejecutar instrucciones **solo si se cumple una condición**.
```python
if condicion:
    # código si la condición es verdadera
else:
    # código si la condición es falsa
```

````python
# Ejemplo simple

edad = int(input("Cual es tu edad? "))

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
````

## Operadores de comparación
- `==` igual a
- `!=` distinto de
- `>` mayor que
- `<` menor que
- `>=` mayor o igual
- `<=` menor o igual

Vamos a probar algunos:

````python
# Prueba de operadores
a = 5
b = 8
print(a == b)  # False
print(a != b)  # True
print(a < b)   # True
````

## Estructura `elif`: múltiples condiciones
Permite verificar varias opciones sin usar muchos `if` anidados:

````python


nota = 6

if nota >= 9:                 # 9 y 10
    print("Excelente")
elif nota >= 7:               #7 y 8
    print("Aprobado")
else:                         # > 7
    print("Desaprobado")
````

## Operadores lógicos: `and`, `or`, `not`
- `and`: ambas condiciones deben ser verdaderas
- `or`: al menos una debe ser verdadera
- `not`: invierte el valor (True ↔ False)

````python
usuario = input("Ingrese su usuario")
clave = int(input("Ingrese su contraseña "))

if usuario == "admin" and clave == 1234:
    print("Acceso concedido")
else:
    print("Acceso denegado")
````
