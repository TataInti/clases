# Clase 6 — Bucles: repetir para aprender

En las clases anteriores implementamos el árbol de decisión con `if` / `elif` / `else`. Funcionó perfecto para **una** fruta.  
Pero KNN no clasifica una fruta — compara la fruta nueva con **cada una** de las frutas del dataset.

Hacer eso a mano sería escribir el mismo código decenas de veces. Para eso existen los **bucles**: estructuras que repiten una acción sobre un conjunto de datos.

---

## 1. El problema que resuelven los bucles

Supongamos que tenemos 5 frutas en el dataset y queremos calcular la distancia de cada una a una fruta nueva.

Sin bucle, el código quedaría así:

````python
# Sin bucle: código repetitivo — cada línea hace lo mismo con datos distintos

nueva_fruta_peso = 175   # gramos

# Calculamos la distancia (diferencia de peso) para cada fruta del dataset
distancia_1 = abs(nueva_fruta_peso - 182)   # fruta 1
distancia_2 = abs(nueva_fruta_peso - 143)   # fruta 2
distancia_3 = abs(nueva_fruta_peso - 170)   # fruta 3
distancia_4 = abs(nueva_fruta_peso - 130)   # fruta 4
distancia_5 = abs(nueva_fruta_peso - 160)   # fruta 5

print(distancia_1, distancia_2, distancia_3, distancia_4, distancia_5)

# Problemas con este enfoque:
# 1. Si el dataset tiene 1000 frutas, necesitamos 1000 líneas de código
# 2. Si el dataset cambia, hay que modificar cada línea por separado
# 3. Es imposible de mantener
print("\nEste código no escala. Si el dataset crece, el código explota.")
````

---

## 2. El bucle `for`: recorrer una colección

El bucle `for` ejecuta un bloque de código **una vez por cada elemento** de una colección.

```python
for elemento in coleccion:
    # código que se ejecuta con cada elemento
```

| Pseudocódigo (Clase 3) | Python |
|---|---|
| `PARA CADA fruta EN el dataset:` | `for fruta in dataset:` |
| `    calcular distancia` | `    distancia = abs(nueva - fruta)` |

> 💡 El `for` de Python corresponde exactamente al "PARA CADA" que escribiste en el pseudocódigo del KNN en la Clase 3.

````python
# El mismo cálculo del bloque anterior — ahora con un bucle for

nueva_fruta_peso = 175   # gramos

pesos_dataset = [182, 143, 170, 130, 160]   # las 5 frutas del dataset

# El bucle recorre la lista y ejecuta el cálculo para CADA peso
for peso_fruta in pesos_dataset:
    distancia = abs(nueva_fruta_peso - peso_fruta)   # diferencia absoluta
    print(f"Distancia a fruta con peso {peso_fruta}g : {distancia}g")

# 5 líneas en lugar de 5 bloques repetidos
# Si el dataset tuviera 10.000 frutas, este mismo código funcionaría igual
````

````python
# El bucle for también funciona con range() para repetir N veces
# (útil cuando no tenemos una lista, sino un número de repeticiones)

print("Contando del 1 al 5:")
for i in range(1, 6):   # range(1, 6) genera los números 1, 2, 3, 4, 5
    print(f"  iteración {i}")

# range(n)       → 0, 1, 2, ..., n-1
# range(a, b)    → a, a+1, ..., b-1
# range(a, b, s) → a, a+s, a+2s, ..., hasta b
````

## 3. El bucle `while`: repetir mientras se cumpla una condición

El bucle `while` ejecuta el bloque de código **mientras** una condición sea `True`.

```python
while condicion:
    # código que se ejecuta mientras la condición sea True
```

| Cuándo usar `for` | Cuándo usar `while` |
|---|---|
| Cuando sabés de antemano cuántas veces vas a iterar | Cuando repetís hasta que algo cambie |
| Recorrer todos los elementos de una lista | Buscar hasta encontrar un resultado |
| Iterar sobre un dataset completo | Entrenar hasta que el error sea suficientemente bajo |

> 💡 El entrenamiento de modelos de ML usa `while` internamente: "seguir ajustando los parámetros *mientras* el error sea mayor al umbral".

````python
# Ejemplo de while: buscar hasta encontrar algo

# Simulamos el proceso de ajuste de un modelo:
# seguimos iterando mientras el error sea demasiado grande

error_actual = 100.0   # error inicial (grande)
umbral       = 5.0     # queremos error menor a este valor
iteracion    = 0

while error_actual > umbral:
    error_actual = error_actual * 0.6   # cada iteración, el error se reduce un 40%
    iteracion   += 1                    # contamos cuántas iteraciones necesitamos
    print(f"Iteración {iteracion}: error = {error_actual:.2f}")

print(f"\nConvergió en {iteracion} iteraciones. Error final: {error_actual:.2f}")
print("Esto es lo que hace el algoritmo de gradient descent cuando entrena un modelo.")
````

## 4. Combinar bucles con condicionales

La potencia real aparece cuando combinamos `for` con `if`. Eso permite iterar sobre datos y tomar decisiones dentro del loop — exactamente lo que hace KNN.

````python
# Dataset de frutas: lista de diccionarios (lo veremos más en la Clase 7)
# Por ahora alcanza saber que cada fruta tiene peso, color y clase

dataset = [
    {"peso": 182, "color": "rojo",    "clase": "manzana"},
    {"peso": 143, "color": "naranja", "clase": "naranja"},
    {"peso": 170, "color": "verde",   "clase": "manzana"},
    {"peso": 130, "color": "naranja", "clase": "naranja"},
    {"peso": 160, "color": "rojo",    "clase": "manzana"},
    {"peso": 145, "color": "amarillo","clase": "limón"},
]

# Clasificar CADA fruta del dataset usando el árbol de decisión de la Clase 5
print("Clasificación de cada fruta del dataset:")
print("-" * 50)

for fruta in dataset:   # recorremos cada elemento de la lista
    peso  = fruta["peso"]
    color = fruta["color"]

    # El árbol de decisión de la Clase 5 adentro del loop
    if peso > 150 and (color == "rojo" or color == "verde"):
        prediccion = "manzana"
    elif peso <= 160 and color == "naranja":
        prediccion = "naranja"
    else:
        prediccion = "otro"

    # Verificamos si la predicción coincide con la clase real
    acierto = "✅" if prediccion == fruta["clase"] else "❌"
    print(f"{acierto} peso={peso}g color={color:8s} | predicho={prediccion:8s} | real={fruta['clase']}")
````

## 5. Acumular resultados: variables fuera del bucle

Una técnica muy común es inicializar una variable *antes* del loop y actualizarla *dentro* del loop para acumular un resultado.

````python
# Encontrar la fruta más cercana a una nueva (KNN con k=1)
# Implementación del pseudocódigo que escribimos en la Clase 3

dataset = [
    {"peso": 182, "color": "rojo",    "clase": "manzana"},
    {"peso": 143, "color": "naranja", "clase": "naranja"},
    {"peso": 170, "color": "verde",   "clase": "manzana"},
    {"peso": 130, "color": "naranja", "clase": "naranja"},
    {"peso": 160, "color": "rojo",    "clase": "manzana"},
]

nueva_fruta = {"peso": 175, "color": "rojo"}   # la fruta a clasificar

# --- Paso 1: Inicializar variables antes del loop ---
menor_distancia = float("inf")   # infinito: cualquier distancia va a ser menor
vecino_mas_cercano = None        # todavía no encontramos ningún vecino

# --- Paso 2: Recorrer el dataset calculando distancias ---
for fruta_conocida in dataset:
    # Distancia = diferencia absoluta de pesos
    distancia = abs(nueva_fruta["peso"] - fruta_conocida["peso"])

    # ¿Esta fruta es más cercana que la más cercana actual?
    if distancia < menor_distancia:
        menor_distancia    = distancia          # actualizamos el mínimo
        vecino_mas_cercano = fruta_conocida     # guardamos cuál es

# --- Paso 3: Devolver la clase del vecino más cercano ---
print(f"Fruta nueva     : peso={nueva_fruta['peso']}g, color={nueva_fruta['color']}")
print(f"Vecino más cercano: peso={vecino_mas_cercano['peso']}g, clase={vecino_mas_cercano['clase']}")
print(f"Distancia       : {menor_distancia}g")
print(f"\nPredicción KNN (k=1): {vecino_mas_cercano['clase']}")
print("\nEsto es exactamente lo que hace scikit-learn con KNN(n_neighbors=1).")
````
