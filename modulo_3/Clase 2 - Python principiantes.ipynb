### Teoría
**1. Operaciones Matemáticas en Python**

En esta clase, exploraremos cómo Python puede funcionar como una calculadora avanzada mediante operadores aritméticos.

Las operaciones numéricas son un pilar fundamental de la programación, utilizadas en disciplinas como la ciencia, la ingeniería, las finanzas y el análisis de datos.

Python nos permite manipular números de manera sencilla y eficiente con los siguientes operadores:

- Suma (`+`): Suma dos números. Ejemplo: `3 + 2` da como resultado `5`.
- Resta (`-`): Resta un número de otro. Ejemplo: `5 - 2` da como resultado `3`.
- Multiplicación (`*`): Multiplica dos números. Ejemplo: `4 * 3` da como resultado `12`.
- División (`/`): Divide un número entre otro, devolviendo un resultado con decimales (tipo float). Ejemplo: `10 / 3` da como resultado aproximadamente `3.333`.
- División entera (`//`): Divide y redondea hacia abajo al número entero más cercano (tipo int). Ejemplo: `10 // 3` da como resultado `3`.
- Resto o Módulo (`%`): Devuelve el resto de una división. Ejemplo: `10 % 3` da como resultado `1` (porque 3 * 3 = 9 y sobra 1).
- Potencia (`**`): Eleva un número a una potencia. Ejemplo: `2 ** 3` da como resultado `8` (2 al cubo).


Un aspecto importante es la precedencia de operadores, que determina el orden en que Python realiza los cálculos. Multiplicación, división, división entera, resto y potencia tienen mayor precedencia que suma y resta. Puedes usar paréntesis () para forzar un orden específico. Por ejemplo:

- `2 + 3 * 4 = 14` (porque `3 * 4` se calcula primero).
- `(2 + 3) * 4 = 20` (los paréntesis obligan a sumar primero).

Estas operaciones son esenciales para resolver problemas numéricos y construir algoritmos básicos, desde calcular promedios hasta modelar datos complejos.

````python
# 2 + 3 * 4 da 14 (porque 3 * 4 se calcula primero)
print("Primer ejemplo '2 + 3 * 4' es igual a = ", 2 + 3 * 4)

# (2 + 3) * 4 da 20 (los paréntesis obligan a sumar primero)
print("Segundo ejemplo '(2 + 3) * 4' es igual a = ", (2 + 3) * 4)
````

**2. Operadores de Comparación: Evaluando Condiciones**

Además de realizar cálculos, en programación necesitamos comparar valores para evaluar condiciones o tomar decisiones. Los operadores de comparación nos permiten hacer esto y devuelven un resultado booleano: True (verdadero) o False (falso). Estos operadores son la base para construir lógica en los programas.

Los operadores de comparación en Python son:

  - Igual a (`==`): Comprueba si dos valores son iguales. Ejemplo: `5 == 5` devuelve `True`.
  - Diferente de (`!=`): Comprueba si dos valores no son iguales. Ejemplo: `5 != 3` devuelve `True`.
  - Mayor que (`>`): Determina si un valor es mayor que otro. Ejemplo: `7 > 3` devuelve `True`.
  - Menor que (`<`): Determina si un valor es menor que otro. Ejemplo: `2 < 5` devuelve `True`.
  - Mayor o igual que (`>=`): Comprueba si un valor es mayor o igual a otro. Ejemplo: `5 >= 5` devuelve `True`.
  - Menor o igual que (`<=`): Comprueba si un valor es menor o igual a otro. Ejemplo: `3 <= 4` devuelve `True`.

Estos operadores se utilizan para comparar números, texto (en algunos casos) y otros datos. Por ejemplo, podrías usarlos para verificar si un estudiante aprobó un curso comparando su nota con un umbral mínimo, o para validar datos ingresados por un usuario.

````python
# Usando condiciones para avanzar en un flujo de trabajo

# Solicitamos edad
edad_usuario = int(input("Cual es tu edad "))

# Flujo de validación básico
if edad_usuario >= 18:              # realizamos la comparación con el primer parametro
  print("Puede continuar")          # Mostramos al usuario el resultado de validación positivo
  zona = input("Donde vives")       # Podemos solicitar otro ingreso de datos
  print("Tienes ", edad_usuario, "y vives en", zona)
else:
  print("No puede continuar")       # Mostramos al usuario el resultado negativo de validación
````

**3. Interacción con el Usuario: Uso de input()**

En muchos programas, es necesario interactuar con el usuario para obtener datos en tiempo real. Python proporciona la función input(), que muestra un mensaje en la consola y espera que el usuario ingrese un valor. El dato ingresado se devuelve como una cadena de texto (tipo str), lo que significa que si queremos trabajar con números, debemos convertirlo a un tipo numérico.

Por ejemplo:

`nombre = input("¿Cuál es tu nombre? ")` muestra el mensaje "¿Cuál es tu nombre?" y guarda la respuesta del usuario como texto en la variable `nombre`.
Si queremos un número, usamos `input()` junto con una conversión: `edad = int(input("Ingresa tu edad: "))` convierte el texto ingresado en un número entero (tipo int).

Es crucial guiar al usuario con mensajes claros para evitar entradas inválidas. Si el usuario ingresa algo que no puede convertirse (por ejemplo, letras cuando se espera un número), Python generará un error. Más adelante aprenderemos a manejar estos errores, pero por ahora nos enfocaremos en conversiones básicas.

````python
# Solicitar datos al usuario
nombre = input("¿Cuál es tu nombre? ")
edad = int(input("Ingresa tu edad: "))

# Mostrar los datos ingresados
print("Hola,", nombre)
print("Tienes", edad, "años.")
````

**4. Conversión de Tipos de Datos: El Concepto de "Casteo"**

El proceso de convertir un dato de un tipo a otro se conoce como "casteo" o conversión de tipos. Como mencionamos, `input()` siempre devuelve una cadena de texto (`str`), pero a menudo necesitamos números para realizar cálculos. Para ello, usamos funciones de conversión como:

 - `int()`: Convierte una cadena de texto a un número entero. Ejemplo: `int("5")` devuelve `5` como tipo int.
 - `float()`: Convierte una cadena de texto a un número decimal. Ejemplo: `float("3.14")` devuelve `3.14` como tipo float.
 - `str()`: Convierte un número u otro dato a texto. Ejemplo: `str(42)` devuelve `"42"` como tipo str.

El "casteo" es esencial para manejar datos ingresados por el usuario de manera adecuada. Por ejemplo, si pides dos números para sumarlos, debes convertirlos con `int()` o `float()`; de lo contrario, Python intentará concatenar las cadenas de texto en lugar de sumar (por ejemplo, "5" + "3" da "53" como texto, no 8).

Un error común es intentar convertir algo inválido, como `int("abc")`, lo que causa una excepción. Por ahora, asumiremos que los usuarios ingresan datos válidos.

````python
# Ejemplo, "5" + "3" da "53" como texto, no 8).
a = str(5)
b = str(3)

# ahora realizamos una operación de "suma" de texto o concatenado
suma_texto = a + b
print(suma_texto)
````

**5. Introducción a las Estructuras Condicionales: `if`, `elif` y `else`**

Un aspecto central de la programación es la capacidad de tomar decisiones basadas en condiciones. Las estructuras condicionales nos permiten ejecutar diferentes bloques de código dependiendo de si una condición es verdadera (`True`) o falsa (`False`). En Python, usamos las palabras clave `if`, `elif` y `else` para construir estas decisiones.

````python
# La sintaxis en Python depende de la indentación (sangría), que es el espacio al inicio de las líneas de código dentro de un bloque.
# Python usa indentación para agrupar instrucciones bajo una misma condición.

edad = 40

if edad >= 18:      # Evalúa una condición; si es `True`, ejecuta el bloque de código asociado
  print("Eres mayor de edad")
elif edad >= 13:    # Se usa después de un if para evaluar condiciones adicionales si la primera no se cumple
  print("Eres adolescente")
else:               # Ejecuta un bloque de código si ninguna de las condiciones anteriores es verdadera
  print("Eres niño")
````



En este caso, el resultado será "Eres adolescente". Las estructuras condicionales combinadas con operadores de comparación permiten que nuestros programas sean dinámicos y respondan de manera inteligente a diferentes situaciones.
