# Clase 4: Repetición con Bucles en Python
En esta clase aprenderás a usar **ciclos** para repetir acciones en tu código. Vamos a trabajar con `while` y `for`, dos estructuras muy utilizadas para automatizar tareas repetitivas.

## ¿Qué es un ciclo?
Un **ciclo** (o bucle) permite ejecutar un bloque de código varias veces, dependiendo de una **condición** (`while`) o de una **secuencia** (`for`).

### ¿Qué es un bucle y por qué es útil?

Un bucle es una estructura de control que permite repetir instrucciones varias veces de forma automática. Es fundamental en programación porque:

- Automatiza tareas repetitivas.

- Permite procesar conjuntos de datos (como listas, archivos o entradas del usuario).

- Hace el código más corto, limpio y eficiente.

.

**Ejemplo del mundo real:**

Contar del 1 al 10 escribiendo print(1), print(2)... hasta print(10) sería ineficiente.

```
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
...
```

Con un bucle, lo hacemos en 2 líneas:

````python
for numero in range(1,11):
  print(numero)
````

### Tipos de bucles en Python

**1. WHILE (MIENTRAS): ciclo controlado por condición**

Se repite mientras la condición sea True.

Puede usarse cuando no sabemos cuántas veces se va a repetir.

```
x = 1
while x <= 3:
    print("x vale", x)
    x += 1
```
*⚠️ Importante: hay que asegurarse de que la condición cambie dentro del bucle para evitar un ciclo infinito.*

````python
x = 1
while x <= 3:
    print("x vale", x)
    x += 1
````

===

**2. FOR (PARA/POR): ciclo sobre una secuencia**

Se usa para recorrer listas, rangos, cadenas, etc.

Ideal cuando sabemos cuántas veces queremos repetir algo.

```
for i in range(5):  # del 0 al 4
    print("i vale", i)
```

También podemos usarlo con listas:
```
colores = ["rojo", "verde", "azul"]
for color in colores:
    print(color)
```

````python
# Recorrer rangos
for i in range(5):  # del 0 al 4
    print("i vale", i)
````

````python
# Recorrer listas
lista = ["rojo", "verde", "azul"]

for color in lista:
    print(color)
````

===


### Estructura interna del bucle
Ambos bucles (for y while) tienen una estructura de bloque, lo que significa que el código a repetir debe ir indentado (con tabulación o espacios).

---

## Bucle `while`
El bucle `while` se ejecuta **mientras** una condición sea verdadera.
```python
contador = 1
while contador <= 5:
    print("Repetición", contador)
    contador += 1
```

## Bucle `for`
El bucle `for` recorre una **secuencia** como una lista, o un rango de números.
```python
for i in range(1, 6):
    print("Repetición", i)
```

## Ejemplo: Recorrer una lista
```python
frutas = ["manzana", "banana", "naranja"]
for fruta in frutas:
    print("Me gusta la", fruta)
```

## Instrucción `break`
Sirve para **interrumpir** un ciclo antes de que termine.
```python
while True:
    entrada = input("Escribe 'salir' para terminar: ")
    if entrada == "salir":
        break
    print("Ingresaste esta palabra:", entrada)
```

## Actividad 1: Contador con `while`
Escribe un bucle que imprima los números del 1 al 10.

````python
# Completa el codigo faltante
contador = 1
while contador <= 10:
  print(contador)
  contador +=1
````

### Solucion

````python
# Solución (oculta)
contador = 1
while contador <= 10:
    print(contador)
    contador += 1
````

## Actividad 3: Contar letras de una palabra
Pide al usuario una palabra y cuenta cuántas letras contiene usando un ciclo `for`.

````python
palabra = input("Escribe una palabra: ")
contador = 0
for letra in palabra:
  contador += 1
  print("La palabra tiene", contador, "letras")
````

### Solucion

````python
# Solución (oculta)
palabra = input("Escribe una palabra: ")
contador = 0
for letra in palabra:
    contador += 1
print("La palabra tiene", contador, "letras.")
````
