## 1. Del pseudocódigo al código Python

Recordá el pseudocódigo de la Clase 3:

```
Paso 2: ¿El peso es mayor a 150 gramos?
            SI  →  ir al Paso 3
            NO  →  ir al Paso 4
```

En Python, ese "SI / NO" se escribe así:

```python
if peso > 150:
    # código que se ejecuta si la condición es True
else:
    # código que se ejecuta si la condición es False
```

| Pseudocódigo | Python |
|---|---|
| `SI condición → hacer algo` | `if condicion:` |
| `SI NO → hacer otra cosa` | `else:` |
| `SI condición A → ...` / `SI NO, y condición B → ...` | `if` / `elif` |

> 💡 La **indentación** (sangría) es obligatoria en Python. El código que va dentro de un `if` debe estar desplazado 4 espacios (o 1 Tab). Python usa eso para saber qué código pertenece a qué bloque.

````python
# El bloque más simple: if / else

peso = 182   # gramos

if peso > 150:
    print("La fruta es pesada (posiblemente manzana)")   # se ejecuta si True
else:
    print("La fruta es liviana (posiblemente naranja)")  # se ejecuta si False

# Cambiá peso a 130 y volvé a ejecutar — el resultado cambia
````

## 2. El árbol de decisión completo con `if` / `elif` / `else`

Cuando hay más de dos caminos posibles, usamos `elif` ("else if": si no lo anterior, verificar esto otro).

```
Árbol del Módulo 2
       ┌──────────────┐
       │  peso > 150  │
       └───┬──────┬───┘
          SÍ     NO
    ┌──────┴──┐  ┌───┴──────────────┐
    │¿rojo o  │  │ ¿color naranja?  │
    │ verde?  │  └──┬───────┬───────┘
    └──┬───┬──┘    SÍ      NO
      SÍ  NO      🍊     🤷
      🍎  🤷
```

Ese árbol tiene 4 ramas. `if` / `elif` / `elif` / `else` las cubre todas.

````python
# Árbol de decisión del Módulo 2 — ahora en código Python real

peso  = 182
color = "rojo"

# Primera condición: peso
if peso > 150 and (color == "rojo" or color == "verde"):
    # Rama: pesada Y color típico de manzana
    clasificacion = "manzana"

elif peso <= 150 and color == "naranja":
    # Rama: liviana Y color naranja
    clasificacion = "naranja"

elif peso > 150 and color == "naranja":
    # Caso borde: naranja muy grande
    clasificacion = "posiblemente naranja grande"

else:
    # Ninguna regla coincide — el modelo no sabe
    clasificacion = "desconocida"

print(f"Peso: {peso}g | Color: {color} → {clasificacion}")

# Probá con estos valores:
# peso=143, color='naranja'  → naranja
# peso=165, color='verde'    → manzana
# peso=120, color='amarillo' → desconocida
````

## 3. El error más común: la indentación

Python usa la sangría para definir los bloques de código. Un error de indentación rompe el programa.

````python
# ✅ Correcto: el código dentro del if está indentado
peso = 182

if peso > 150:
    print("Fruta pesada")    # 4 espacios de sangría
    print("Puede ser manzana")  # mismo nivel de sangría

print("Esta línea siempre se ejecuta, sin importar el if")
````

````python
# Lo que pasa cuando usamos variables de la Clase 4
# (las expresiones booleanas que ya calculamos)

peso  = 143
color = "naranja"

# Podemos calcular primero la condición y guardarla en una variable
es_liviana      = peso <= 150
es_color_naranja = color == "naranja"

# Y después usarla en el if
if es_liviana and es_color_naranja:
    print("🍊 Es una naranja")
else:
    print("No es una naranja típica")

# Esto produce el mismo resultado que escribir la condición directamente en el if
# pero es más legible cuando las condiciones son complejas
````

## 4. Condicionales anidados

Un `if` puede contener otro `if` adentro. Eso se llama **anidamiento** y corresponde exactamente a los nodos internos del árbol de decisión.

> 💡 Los árboles de decisión de scikit-learn funcionan exactamente así internamente: un `if` dentro de otro. La diferencia es que los parámetros los aprende automáticamente del dataset.

````python
# Condicionales anidados: un if dentro de otro
# Corresponde a los nodos internos del árbol de decisión

peso  = 182
color = "verde"    # manzana Granny Smith

if peso > 150:
    # Primer nodo superado: fruta pesada
    # Ahora evaluamos el segundo nodo (color)
    if color == "rojo" or color == "verde":
        print("🍎 Es una manzana (pesada + color típico)")
    else:
        print("Fruta pesada pero color desconocido")
else:
    # Primer nodo no superado: fruta liviana
    if color == "naranja":
        print("🍊 Es una naranja (liviana + naranja)")
    else:
        print("Fruta liviana pero color desconocido")

# Fijate que la estructura del código replica exactamente el diagrama del árbol
````

## ✅ Resumen

| Estructura | Cuándo usarla |
|---|---|
| `if condicion:` | Cuando querés ejecutar código solo si algo es verdadero |
| `if ... else:` | Cuando hay dos caminos posibles |
| `if ... elif ... else:` | Cuando hay más de dos caminos (árbol con múltiples ramas) |
| Condicionales anidados | Cuando una decisión depende de otra anterior |

**Conexión con ML:**
- El árbol de decisión de scikit-learn es una cadena de `if` / `elif` cuyas condiciones fueron *aprendidas* del dataset
